> ## Prelude:  Density matrices
<comment>

Density matrices are a way to describe quantum systems. The easiest way to understand them is to consider a 2 level system whose states are given by $|0 \rangle $ and $|1 \rangle$. Now let us say that a source produces these states with probability p and 1-p respectively (note that this is classical probability, not quantum probablity) then we can write the state as:
$$
\rho = p |0 \rangle \langle 0 | + (1-p) |1 \rangle \langle 1|
$$
where $\rho$ is a density matrix, let us observe some properties of this density matrix from this form that we will generalise to all density matrices:
* $\rho$ is hermitian with tr($\rho$) = 1
* $\rho$ is positive semi-definite i.e has $\geq 0$ eigenvalues, this is simply stating that the probabilties p and (1-p) have to be positive semi definitie
* $\langle A \rangle$ = Tr$(\rho A)$ (use the cyclicity of trace to prove this)
* Pure density matrices are given by $tr(\rho^2) = 1$, they are density matrices that can be represnted as $| \psi \rangle \langle \psi |$ in some basis

There is a time evolution equation for density matrices that we can get by proceeding in the following manner: 
$$
i \hbar | \dot{\psi} \rangle = \hat{H} | \psi \rangle\\
\dot{\rho} = | \dot{\psi} \rangle \langle \psi | + | \psi \rangle \langle \dot{\psi} | \\
\implies \dot{\rho} = \frac{ -i}{\hbar} [ \hat{H}, \rho]
$$


## ENZ (Epsilon near zero) materials
### How can Epsilon get Near Zero ?
<a id='sec1'></a>
Working with complex refractive index and dielectric constant we have:
$$
n = n_r + i n_i \\
ϵ = ϵ_r + i ϵ_i
$$

using the same fundamental relation $n^2 = \epsilon_0 ϵ$

$$
ϵ_r = ϵ_0 (n_r^2 - n_i^2)\\
ϵ_i = ϵ_0(2 n_r n_i)\\
n_r = \sqrt{\frac{| \epsilon| + \epsilon_r}{2 \epsilon_0}}\\
n_i = \sqrt{\frac{| \epsilon| - \epsilon_r}{2 \epsilon_0}}
$$

These relations will be useful for us later on when we write things out as linear and non linear terms.

$$
D = ϵ_0 E + P = \epsilon_0 (1 + \chi) E = \epsilon_0 (1 + \chi^{(1)}(w) + 3 \chi^{(3)} (w)|E|^2) E = \epsilon E\\
\epsilon = \epsilon^{(1)} + \epsilon^{(NL)}\\
\implies \epsilon^{(NL)}_r = 3 \epsilon_0 \chi_r^{(3)} |E|^2
, \epsilon^{(NL)}_i = 3 \epsilon_0 \chi_i^{(3)} |E|^2
$$
similary we write $n = n^{(1)} + n^{(NL)} = n^{(1)} + n^{(2)} I$ (intensity dependent refractive index, $I = 2 \epsilon_0 n^{(1)} c |E|^2$). Using all the relations above we simplify to obtain:
$$
\epsilon^{(NL)}_r = \epsilon_0 [ 2(n^{(1)}_r n^{(NL)}_r - n^{(1)}_i n^{(NL)}_i) + (n^{(NL)}_r)^2 + (n^{(NL)}_i)^2 ]
$$
taking the assumption that $n^{(NL)}_i << n^{(1)}_i$ and $n^{(NL)}_r << n^{(1)}_r$ we get,
$$
3 \epsilon_0 \chi_r^{(3)} |E|^2 = 4 \epsilon_0 n^{(1)} c |E|^2(n^{(1)}_r n^{(2)}_r - n^{(1)}_i n^{(2)}_i)\\
\text{similarly, } 3 \epsilon_0 \chi_i^{(3)} |E|^2 = 4 \epsilon_0 n^{(1)} c |E|^2(n^{(1)}_r n^{(2)}_r + n^{(1)}_i n^{(2)}_i)
$$
inverting these relations to get refractive index as a function of susceptibiity:
$$
n^{(2)}_r = \frac{3}{4\epsilon_0 c( (n_r^{(1)})^2 + (n_i^{(1)})^2 }[ \chi_r^{(3)} + \frac{n^{(1)}_i}{n^{(1)}_r} \chi_i^{(3)}] \\
n^{(2)}_i = \frac{3}{4\epsilon_0 c( (n_r^{(1)})^2 + (n_i^{(1)})^2 }[ \chi_r^{(3)} - \frac{n^{(1)}_i}{n^{(1)}_r} \chi_i^{(3)}] 
$$

having derived these relations we can now describe what we mean by ENZ. For a material with complex dielectric constant (which is dependent on the frequency $ \omega$) for a particular $\omega$, $\epsilon^{(1)}_r = 0$ while imaginary part remains nonzero. Note that this does not mean your refractive index goes to zero as can be seen from the formula we have written for refractive index.

> ## Two-Level Atomic System

<a id='sec2'></a>
<comment>

We begin with an atomic system that has two energy eigenstates a and b with energies $E_a = {\hbar}\omega_a$ and $E_b={\hbar}{\omega_b}$ respectively.<break>

The Hamiltonian for this system is given by $\hat{H} = \hat{H}_0 + \hat{V}(t)$, where $\hat{H}_0$ is the Hamiltonian of the system in absence of extrernal disturbances and $\hat{V}(t)$ represents the interaction of the atomic system with electromagnetic radiation. From the energy levels, we can write $\hat{H}_{0,nm} = \delta_{nm}E_n$, since it will be diagonal. In terms of the atomic dipole moment, the interaction term can be described as $\hat{V}(t) = -\hat{\mu}{\cdot}E(t)$ <break>

$$
\implies \hat{H}_{nm} = \delta_{nm}E_n -\hat{\mu}_{nm}E(t)
$$

We can now write the density matrix equations of this system as 

$$
\dot{\rho}_{nm} = \frac{-i}{\hbar}[\hat{H}, \hat{\rho}]_{nm} = \frac{-i}{\hbar}{\sum_{\nu}H_{n{\nu}}{\rho}_{{\nu}{m}} - {\rho}_{n{\nu}}{H}_{{\nu}{m}}}
$$

Substituting $\hat{H}_{nm} = \delta_{n{\nu}}E_n + \hat{V}_{nm}$, we get

$$
\dot{\rho}_{nm} = \frac{-i}{\hbar}{\sum_{\nu}V_{n{\nu}}{\rho}_{{\nu}{m}} - {\rho}_{n{\nu}}{V}_{{\nu}{m}}} - \frac{i}{\hbar}(E_n - E_m)\rho_{nm} = \frac{-i}{\hbar}{\sum_{\nu}(V_{n{\nu}}{\rho}_{{\nu}{m}} - {\rho}_{n{\nu}}{V}_{{\nu}{m}})} - i\omega_{nm}\rho_{nm}
$$

From the properties of the dipole moment, we also have $V_{aa} = V_{bb} = 0$ since $\mu_{aa} = \mu_{bb} = 0$ (from parity arguments). So we separate the density matrix equation into diagonal and off diagonal components:

$$
\begin{aligned}
\dot{\rho}_{aa} &= \frac{-i}{\hbar}({V_{ab}{\rho}_{ba} - {\rho}_{ab}{V}_{ba}}) \\
\dot{\rho}_{bb} &= \frac{i}{\hbar}({V_{ab}{\rho}_{ba} - {\rho}_{ab}{V}_{ba}}) \\
\dot{\rho}_{ab} &= \frac{-i}{\hbar}{\sum_{\nu}(V_{a{\nu}}{\rho}_{{\nu}{b}} - {\rho}_{a{\nu}}{V}_{{\nu}{b}})} - i\omega_{ab}\rho_{ab} = \frac{i}{h}V_{ab}(\rho_{aa} - \rho_{bb})- i\omega_{ab}\rho_{ab} 
\end{aligned}
$$

$\Delta \rho = \rho_{11} − \rho_{22}$ is the population difference between the lower and upper levels. The off-diagonal terms $\rho_{21} \text{and } \rho_{12}$ are sometimes called coherences and
describe the net atomic polarization.

Now for a closed two-level atom, we introduce relaxation times $T_1$ and $T_2$, where $T_1$ is the lifetime of state b in decay from b to a and $T_2$ is the dephashing time for the dipole moment (This represents the time for loss of coherence between the 2 states (a purely quantum mechanical effect which decays to it's classical value and is measured by the off diagonal terms of density matrix), it is caused by frequency of light close to the energy difference between the 2 levels). The equations are then modified by introducing $\frac{\rho_{bb}}{T_1}$ (which represents the rate of decay from state b) in the expression for $\dot{\rho}_{aa}$ and $\gamma_{ba} = \frac{1}{T_2}$ (as dipole dephasing amplifies the net atomic polarization) in the expression for $\dot{\rho}_{ab}$.

$$
\begin{aligned}
\dot{\rho}_{aa} &= \frac{\rho_{bb}}{T_1} - \frac{i}{\hbar}({V_{ab}{\rho}_{ba} - {\rho}_{ab}{V}_{ba}}) \\
\dot{\rho}_{bb} &= -\frac{\rho_{bb}}{T_1} + \frac{i}{\hbar}({V_{ab}{\rho}_{ba} - {\rho}_{ab}{V}_{ba}}) \\
\dot{\rho}_{ab} &= \frac{i}{h}V_{ab}(\rho_{aa} - \rho_{bb})- (i\omega_{ab} + \gamma_{ab})\rho_{ab}
\end{aligned}
$$

From density matrix equations, we have $\dot{\rho}_{aa} + \dot{\rho}_{bb} = 0$ and ${\rho}_{aa} + {\rho}_{bb} = 1$ Taking $\Delta\rho = \rho_{aa} - \rho_{bb}$, we get 

$$
\frac{d\Delta{\rho}}{dt} = \frac{2\rho_{bb}}{T_1} - \frac{2i}{\hbar}({V_{ab}{\rho}_{ba} - {\rho}_{ab}{V}_{ba}}) = \frac{\rho_{bb} - \rho_{aa} + 1}{T_1} - \frac{2i}{\hbar}({V_{ab}{\rho}_{ba} - {\rho}_{ab}{V}_{ab}^{*}})
$$

Since $V_{ba} = V_{ab}^{*} = -\hat{\mu}_{ba}E(t)$, we get 

$$
\frac{d\Delta{\rho}}{dt} = \frac{-(\Delta{\rho} - 1)}{T_1} - \frac{2i}{\hbar}({V_{ab}{\rho}_{ba} - {\rho}_{ab}{V}_{ab}^{*}}) = \frac{-(\Delta{\rho} - 1)}{T_1} + \frac{2i\mu_{21}E(t)}{\hbar}({\rho}_{ba} - {\rho}_{ab})
$$

Therefore out final set of equations are:

$$
\begin{align}
\tag{1}
\frac{d{\rho}_{ab}}{dt} = \frac{i}{h}V_{ab}\Delta{\rho}- (i\omega_{ab} + \gamma_{ab})\rho_{ab}
\end{align}
$$
$$
\begin{align}
\tag{2}
\frac{d\Delta{\rho}}{dt} = \frac{-(\Delta{\rho} - 1)}{T_1} + \frac{2i\mu_{21}E(t)}{\hbar}({\rho}_{ba} - {\rho}_{ab})
\end{align}
$$

From these density matrix equations, we write the electric field as $E(t) = E_1e^{-i{\omega_1}t} + E_2e^{-i{\omega_2}t} + c.c$. Under the rotating wave approximation, we will discard terms that have an exponent of $\omega_1 + \omega_2$. We proceed to use perturbation theory, and write $\Delta{\rho}$ and $\rho_{21}$ as

$$
\begin{align}
\tag{3}
\Delta{\rho} = 1 - \delta_{12}e^{-i({\omega_1 - \omega_2})t} - \delta_{12}^{*}e^{i({\omega_1 - \omega_2})t} + ... 
\end{align}
$$
$$
\begin{align}
\tag{4}
\rho_{21} = {{\sigma}_1}{e^{-i{\omega_1}t}} + {{\sigma}_2}{e^{-i{\omega_2}t}} + {{\sigma}_{112}}e^{-i({2\omega_1 - \omega_2})t} + {{\sigma}_{211}}e^{-i({2\omega_2 - \omega_1})t} + ... 
\end{align}
$$

We need to calculate the third order susceptibility. So we proceed by calculating all the coefficients of the perturbative expansion till $\sigma_{112}$. Substituting $(4)$ in $(1)$ and taking all the **first order** terms oscillating at frequency $\omega_1$, we get (taking $\gamma_{21} = \gamma_{12} = \gamma = \frac{1}{T_2}$)

$$
\begin{aligned}
\frac{d}{dt}{{{\sigma}_1}{e^{-i{\omega_1}t}}} &= - (i\omega_{21} + \gamma){{\sigma}_1}{e^{-i{\omega_1}t}} + \frac{i}{h}{\mu_{21}(E_1e^{-i{\omega_1}t} + E_2e^{-i{\omega_2}t} + c.c)}(1 - \delta_{12}e^{-i({\omega_1 - \omega_2})t} - \delta_{12}^{*}e^{i({\omega_1 - \omega_2})t}) \\
-i{\omega_1}{\sigma_1}{e^{-i{\omega_1}t}} &= - (i\omega_{21} + \gamma){{\sigma}_1}{e^{-i{\omega_1}t}} + \frac{i}{h}{\mu_{21}}E_1e^{-i{\omega_1}t} \\
{\sigma_1}(-i{\omega_1} + i\omega_{21} + \gamma) &=  \frac{i}{h}{\mu_{21}}E_1 \\
\sigma_1 &= \frac{{\mu_{21}}E_1{\hbar}^{-1}}{({\omega_{21}} - \omega_{1}) - i\gamma}
\end{aligned}
$$

Similarly, for $\omega_2$, we get

$$
\begin{aligned}
\frac{d}{dt}{{{\sigma}_2}{e^{-i{\omega_2}t}}} &= - (i\omega_{21} + \gamma){{\sigma}_2}{e^{-i{\omega_2}t}} + \frac{i}{h}{\mu_{21}(E_1e^{-i{\omega_1}t} + E_2e^{-i{\omega_2}t} + c.c)}(1 - \delta_{12}e^{-i({\omega_1 - \omega_2})t} - \delta_{12}^{*}e^{i({\omega_1 - \omega_2})t}) \\
-i{\omega_2}{\sigma_1}{e^{-i{\omega_2}t}} &= - (i\omega_{21} + \gamma){{\sigma}_2}{e^{-i{\omega_2}t}} + \frac{i}{h}{\mu_{21}}E_2e^{-i{\omega_2}t} \\
{\sigma_2}(-i{\omega_2} + i\omega_{21} + \gamma) &=  \frac{i}{h}{\mu_{21}}E_2 \\
\sigma_2 &= \frac{{\mu_{21}}E_2{\hbar}^{-1}}{({\omega_{21}} - \omega_{2}) - i\gamma}
\end{aligned}
$$

Next, we substitute $(3)$ in $(2)$ and take the **first order** terms oscillating at frequency $\omega_1 - \omega_2$ to get $\delta_{12}$

$$
\begin{aligned}
\frac{d}{dt}(1 - \delta_{12}e^{-i({\omega_1 - \omega_2})t}) &= \frac{(\delta_{12}e^{-i({\omega_1 - \omega_2})t})}{T_1} + \frac{i\mu_{21}(E_1e^{-i{\omega_1}t} + E_2e^{-i{\omega_2}t} + E_1^{*}e^{i{\omega_1}t} + E_2^{*}e^{i{\omega_2}t})}{\hbar}(({{\sigma}_1}{e^{-i{\omega_1}t}} + {{\sigma}_2}{e^{-i{\omega_2}t}}) - ({{\sigma}_1^{*}}{e^{i{\omega_1}t}} + {{\sigma}_2^{*}}{e^{i{\omega_2}t}})) \\
i(\omega_1 - \omega_2)\delta_{12}e^{-i({\omega_1 - \omega_2})t} &= \frac{\delta_{12}e^{-i({\omega_1 - \omega_2})t}}{T_1} + \frac{i\mu_{21}}{\hbar}(E_2^{*}\sigma_1 - E_1\sigma_2^{*})e^{-i({\omega_1 - \omega_2})t} \\
(i(\omega_1 - \omega_2) - \frac{1}{T_1})\delta_{12} &= \frac{i\mu_{21}}{\hbar}(\frac{{\mu_{21}}E_2E_1{\hbar}^{-1}}{({\omega_{21}} - \omega_{1}) - i\gamma} - \frac{{\mu_{21}}E_1E_2{\hbar}^{-1}}{({\omega_{21}} - \omega_{2}) + i\gamma}) = \frac{i\mu_{21}^{2}E_1E_2}{\hbar^{2}}(\frac{1}{({\omega_{21}} - \omega_{1}) - i\gamma} - \frac{1}{({\omega_{21}} - \omega_{2}) + i\gamma}) \\
(i(\omega_1 - \omega_2) - \frac{1}{T_1})\delta_{12} &= \frac{i\mu_{21}^{2}E_1E_2}{\hbar^{2}}(\frac{\omega_1 - \omega_2 + 2i\gamma}{({\omega_{21}} - \omega_{1} - i\gamma)({\omega_{21}} - \omega_{2} + i\gamma)}) \\
\delta_{12} &= \frac{{\omega_1 - \omega_2 + 2i\gamma}}{(\omega_1 - \omega_2 + \frac{i}{T_1})}(\frac{{\mu_{21}^{2}E_1E_2}/{\hbar^{2}}}{({\omega_{21}} - \omega_{1} - i\gamma)({\omega_{21}} - \omega_{2} + i\gamma)})
\end{aligned}
$$

Finally, we obtain $\sigma_{112}$ by substituting $(4)$ in $(1)$ again and taking terms oscilalting at frequency $2\omega_1 - \omega_2$. SImilar to the derivation of $\sigma_1$, we get

$$
\begin{aligned}
\frac{d}{dt}{{{\sigma}_{112}}{e^{-i{(2\omega_1 - \omega_2)}t}}} &= - (i\omega_{21} + \gamma){{\sigma}_{112}}{e^{-i{(2\omega_1 - \omega_2)}t}} + \frac{i}{h}{\mu_{21}(E_1e^{-i{\omega_1}t} + E_2e^{-i{\omega_2}t} + c.c)}(1 - \delta_{12}e^{-i({\omega_1 - \omega_2})t} - \delta_{12}^{*}e^{i({\omega_1 - \omega_2})t}) \\
-i({2\omega_1 - \omega_2}){{{\sigma}_{112}}{e^{-i{(2\omega_1 - \omega_2)}t}}} &= - (i\omega_{21} + \gamma){{\sigma}_{112}}{e^{-i{(2\omega_1 - \omega_2)}t}} - \frac{i}{h}{\mu_{21}}\delta_{12}E_1e^{-i{(2\omega_1- \omega_2)}t} \\
({-2i\omega_1 + i\omega_2} + i\omega_{21} + \gamma){{\sigma}_{112}} &=  - \frac{i}{h}{\mu_{21}}\delta_{12}E_1 \\
{{\sigma}_{112}} &= \frac{{\mu_{21}}\delta_{12}E_1/{\hbar}}{(\omega_{2} + \omega_{21} -2\omega_1  - i\gamma)} = \frac{{\omega_1 - \omega_2 + 2i\gamma}}{(\omega_1 - \omega_2 + \frac{i}{T_1})}(\frac{{\mu_{21}^{3}E_1^2E_2}/{\hbar^{3}}}{({\omega_{21}} - \omega_{1} - i\gamma)({\omega_{21}} - \omega_{2} + i\gamma)(\omega_{2} + \omega_{21} -2\omega_1  - i\gamma)})
\end{aligned}
$$

Now, we can obtain the third order susceptibility $\chi^{(3)}$ since $P^{(3)}(2\omega_1 - \omega_2) = N\mu_{21}\sigma_{112}e^{-i(2\omega_1 - \omega_2)t} = \epsilon_0\chi^{(3)}E_1^{2}E_2e^{-i(2\omega_1 - \omega_2)t}$, where $N$ is the number density of the two-level system. We make the assumption that $\omega_1$ and $\omega_2$ are sufficiently close that $\Delta{\omega}<<\omega_1, \omega_2$. Therefore, we can approximate both $\omega_1$ and $\omega_2$ by $\bar{\omega} = \frac{\omega_1 + \omega_2}{2}$. Hence,

$$
\begin{aligned}
\chi^{(3)} &= \frac{N\mu_{21}\sigma_{112}}{\epsilon_0E_1^{2}E_2} \\
\chi^{(3)} &\approx \frac{{\Delta\omega + 2i\gamma}}{(\Delta\omega + \frac{i}{T_1})}(\frac{{N\mu_{21}^{4}}/{\epsilon_{0}\hbar^{3}}}{({\omega_{21}} - \bar{\omega} - i\gamma)({\omega_{21}} - \bar{\omega} + i\gamma)(\Delta\omega + \omega_{21} -\bar{\omega}  - i\gamma)})\\
\chi^{(3)} &\approx \frac{{\Delta\omega + 2i\gamma}}{(\Delta\omega + \frac{i}{T_1})}(\frac{{N\mu_{21}^{4}}/{\epsilon_{0}\hbar^{3}}}{({(\omega_{21}} - \bar{\omega})^2 + \gamma^2)^{\frac{3}{2}}})
\end{aligned}
$$

Now we can separate this susceptibility into fast and slow components as follows

$$
\begin{aligned}
\frac{{\Delta\omega + 2i\gamma}}{\Delta\omega + \frac{i}{T_1}} &= \frac{{\Delta\omega + \frac{i}{T_1} + 2i\gamma - \frac{i}{T_1}}}{\Delta\omega + \frac{i}{T_1}} = 1 + \frac{{2i\gamma - \frac{i}{T_1}}}{\Delta\omega + \frac{i}{T_1}} = 1 + \frac{{2i\frac{T_1}{T_2} - i}}{\Delta{\omega}T_1 + i} = 1 + \frac{{2\frac{T_1}{T_2} - 1}}{1 -i\Delta{\omega}T_1} \\
\therefore \chi^{(3)} &\approx (1 + \frac{{2\frac{T_1}{T_2} - 1}}{1 -i\Delta{\omega}T_1})(\frac{{N\mu_{21}^{4}}/{\epsilon_{0}\hbar^{3}}}{({(\omega_{21}} - \bar{\omega})^2 + \gamma^2)^{\frac{3}{2}}}) = \chi_{fast}^{(3)} +  \chi_{slow}^{(3)} 
\end{aligned}
$$


We can see that the non-linear susceptibility has two components, one of which depends on the beat frequency (slow term) and the other doesn't (fast term). The slow component is related to "real" excitation of electrons, where they occupy an energy eigenstate, having a characteristic relaxation decay time $T_1$. The fast component is associated with transitions to virtual states, in which the electrons spend very less time. The slow component is larger than the fast one by a factor of $2\frac{T_1}{T_2}-1$ As the brat frequency increases, the fast term begins to dominate the susceptibility. <break>

The above analysis also applies to the behavior of dielectric systems, even though they are not a two-level system in a strict sense. Instead, they have two bands in which electrons can reside: the valence and conduction band, with a bandgap $E_{gap}$, which may be wider than the optical bandgap. As long as the frequency of light is such that its energy is less than $E_{gap}$, we can approximate the dielectric system by a two-level system, with the energy difference between the two states approximated by $\hbar\bar{\omega}{21}$, which is larger than $E{gap}$. This difference is called the Penn gap. Since the term corresponding to $\bar\omega$ in the denominator of $\chi^{(3)}$ is less then $E_{gap}$, it's value (order of a few $eV^2$) doesn't change much in this approximation.

```python
def sus(delw,t1,t2):
  return abs(1 + (2*t1/t2 - 1)/(1 - delw * t1 * 1j))
def susfast(delw,t1,t2):
  return abs((2 * t1/t2 - 1)/(1 - delw * t1 * 1j))

plt.figure(figsize=(10, 6))
delws = np.logspace(0, 14, 100)
t1, t2 = 1e-9, 1e-12
sussies = sus(delws, t1, t2)
norm = sussies[0]
sussies /= norm

plt.title('Fast and Slow Susceptibilities vs Beats Frequency', fontsize=20)
plt.plot(delws, sussies)
plt.plot(delws, susfast(delws, t1, t2)/norm, c='r', ls='--', label='Fast', lw=2)
plt.axhline(1/norm, c='b', ls='--', label='Slow', lw=2)
plt.axvline(1/t1, c='k', ls='--', label=r'$T1^{-1}$')
plt.axvline(1/t2, c='k', ls='--', label=r'$T2^{-1}$')
plt.xscale('log')
plt.yscale('log')
plt.ylabel('Susceptibility (arbitrary units)', fontsize=15)
plt.xlabel('Beats frequency (Hz)', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=14)
plt.show()
```

> ## Periodic Potential Kronig-Penney Model

If we ignore edge effects, which is quite accurate because of the size of typical crystals, the potential of a crystal is periodic i.e.
$$V(x+p) = V(x+a+b) = V(x)$$
In such potentials, the wave function must be of the form,
$$|\psi(x+p)|^2 = |\psi(x)|^2 \implies \psi(x+p) = e^{ikp}\psi(x)$$

Ref: EE207, 2021, Prof. Bhaskaran Muralidharan

> ## The Bloch Theorem

Let us also make the assumption that the wave function is "circular periodic". Thus, 
$$\psi(x+Np)=\psi(x)e^{ikpN} = \psi (x)$$
Applying the boundary conditions we have
$$e^{ikpN}=1 \equiv e^{i2n\pi}$$
As a result we have
$$k=\pm \frac{2\pi n}{Np}, n = -\frac{N}{2}, .., -1, 0, 1, .., \frac{N}{2}$$
$$\implies k_{max}=\frac{\pi}{p}, k_{min}=-\frac{\pi}{p}$$
Also, notice that we can write the wavefunction in the form of a plane wave modulated by a periodic function, i.e,
$$\psi(r)=e^{ikr}u(r)$$

Ref: EE207, 2021, Prof. Bhaskaran Muralidharan

Suppose that the wavefunction is of the form
$$
\psi=
\begin{cases}
A_b \sin\beta x + B_b\cos\beta x & -b \le x \le 0 \\
A_a \sin\alpha x + B_a\sin\alpha x & 0 \le x \le a
\end{cases}
$$

where $\alpha \equiv \sqrt{2mE/\hbar^2}, \beta \equiv i\sqrt{2m(U_0-E)/\hbar^2}$. Also define $\alpha_0 \equiv \sqrt{2mU_0/\hbar^2}$.

**Boundary Conditions:**
1. $\psi|_{x=0^-} = \psi|_{x=0^+}$
2. $\frac{d\psi}{dx}|_{x=0^-}=\frac{d\psi}{dx}|_{x=0^+}$
3. $\psi|_{x=a} = \psi|_{x=-b}$
4. $\frac{d\psi}{dx}|_{x=a}=\frac{d\psi}{dx}|_{x=-b}$

**This gives us:**
1. $B_a = B_b$
2. $\alpha A_a = \beta A_b$
3. $A_a\sin\alpha a + B_a\cos\alpha a = e^{ik(a+b)}[-A_b\sin\beta b+B_b\cos \beta b]$
4. $\alpha A_a\cos\alpha a - \alpha B_a\sin\alpha a = e^{ik(a+b)}[\beta A_b\cos\beta b+\beta B_b\sin\beta b]$

**Let** $\zeta = E/U_0$. **Substituting and rearranging we get the following result:**
$$\frac{1-2\zeta}{2\sqrt{\zeta(1-\zeta)}}\sin\alpha_0a\sqrt{\zeta}\sinh\alpha_0b\sqrt{1-\zeta}+\cos\alpha_0a\sqrt{\zeta}\cosh\alpha_0b\sqrt{1-\zeta} = \cos k (a+b) \hspace{2em}\text{... when } \hspace{0.5em} 0 < E < U_0$$
$$\frac{1-2\zeta}{2\sqrt{\zeta(\zeta-1)}}\sin\alpha_0a\sqrt{\zeta}\sin\alpha_0b\sqrt{\zeta-1}+\cos\alpha_0a\sqrt{\zeta}\cos\alpha_0b\sqrt{\zeta-1} = \cos k (a+b) \hspace{2em}\text{... when } \hspace{0.5em} E > U_0$$

Solving the above equations, we get the valid values of $E$, the energy of the particle. Notice that the $E$ can take only discrete values because $k$ can take only certain values. 

Ref: EE207, 2021, Prof. Bhaskaran Muralidharan

```python
# write the above equation as f(xi) = cos(k*(a+b))
# solve for xi to get the valid states of the system

def f(xi):
  return (1-2*xi)/(2*np.sqrt(xi*abs(1-xi)))*(np.sin(a*xi**0.5)*np.sin(a*abs(1-xi)**0.5) + np.cos(b*xi**0.5)*np.cos(b*abs(1-xi)**0.5))
 
kp = np.vectorize(f)
kpp = lambda xi: kp(xi)-1 # cos(k*(a+b)) = +1
kpm = lambda xi: kp(xi)+1 # cos(k*(a+b)) = -1

xis = np.linspace(1.01,2,1000)
a, b = 30, 30
fig, ax = plt.subplots(figsize=(10,8))

# solving for intersection points
minus1 = fsolve(kpm, [1.011,1.05,1.19,1.25,1.6])
plus1 = fsolve(kpp, [1.011,1.05,1.13,1.4,1.6])

# plotting
xranges = xis[np.logical_and(kp(xis)<1, kp(xis)>-1)]
plt.title('$f(\\xi)$ vs $\\xi$', fontsize=20)
plt.plot(xis,kp(xis), label='$f(\\xi)$')
plt.plot(xis, np.ones(len(xis)), label='$k=0$')
plt.plot(xis, -np.ones(len(xis)), label='$k=\\pm\\frac{\\pi}{a+b}$')
for i in range(5):
    ax.add_patch(Rectangle((min(minus1[i],plus1[i]),-1),abs(plus1[i]-minus1[i]),2,facecolor='orange',alpha = 0.7))
plt.ylabel(r'$f(\xi)$', fontsize=15)
plt.xlabel(r'$\xi$', fontsize=15);
plt.legend(fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

> ## Group Velocity
The group velocity of a wave is the velocity with which the overall envelope shape of the wave's amplitudes—known as the modulation or envelope of the wave—propagates through space. 

Let us now calculate the group velocity of the wave. We have
$$\psi(x,t) = A e^{ikx-iEt/\hbar} \left[  1 + e^{i\Delta k x - i\Delta E t/\hbar} \right]$$
$$ = A e^{ikx-iEt/\hbar} \left[  1 + e^{i * const.} \right]$$
$$\implies v = \frac{\Delta x}{\Delta t} = \frac{\Delta E}{\hbar \Delta k}$$
$$ a = \frac{\Delta v}{\Delta t} = \frac{1}{\hbar} \frac{d}{dt}\left[ \frac{\Delta E}{\Delta k} \right] \frac{d(\hbar k)}{dt}=\frac{F}{m_*}$$
Notice that the particles appear to follow Newton's Law of Motion, except that we have to use their **effective mass** $m_*$ which can be calculated by the expression
$$\frac{1}{m_*} = \frac{1}{\hbar^2} \frac{d^2E}{dk^2}$$

2. https://en.wikipedia.org/wiki/Group_velocity

$$E = E_c + Ak_1^2 + B(k_2^2 + k_3^2)$$

> ## Carrier motion in non-parabolic band







Till now, we were dealing with potentials which can be accurately approximated using parabolas. This, however, need not be true for every potential. 

Consider the motion of electrons in a band with dispersion $E(k)$, in the presence of a harmonic electric field $E_z(t) = E_1e^{i\omega_1 t} + E_2e^{i\omega_2 t} + \text{c.c.}$

The equation of motion will thus be
$$\frac{dk_z}{dt} = -\frac{e}{\hbar}E_z - \gamma(k_z-k_{z0})$$ Where $\gamma$ is the momentum scattering rate. The solution for the harmonic input field will be $k_z(t) = k_{z0} + \delta k (t)$, where
$$\delta k(t) = \frac{e}{\hbar(i\omega_1-\gamma)}E^{-i\omega_1t} + \frac{e}{\hbar(i\omega_2-\gamma)}E^{-i\omega_2t} + \text{c.c.}$$

The electron velocity in the band $E(k)$ in the direction of the applied electric field is $v_z = \hbar^{-1}\partial E/\partial k_z$, which can be expanded into the power series around $k_0$ as 
$$v_z(t) = v_z(k_{z0}) + \frac{\partial v_z}{\partial k_z} \delta k(t) + \frac{1}{2}\frac{\partial^2 v_z}{\partial k_z^2} \delta k^2(t) + \frac{1}{6}\frac{\partial^3 v_z}{\partial k_z^3} \delta k^3(t) + \cdots$$

Since the system has time reversal symmetry $v(k)=-v(-k)$, the summation over all filled states in the band will produce a net zero velocity for the time independent term. The first remaining term that is linear in $\delta k(t)$ can be represented by introducing the _optical effective momentum dependent mass_
$$m_{t}^{-1} = \hbar^{-1} \frac{\partial v_z}{\partial k_z}$$

> **Deriving the Optical Effective Momentum Dependent Mass**

For a field applied in the z-direction, the velocity associated with a particular momentum state becomes:
$$v_z = v\cos\theta = v\frac{k_z}{\sqrt{k_z^2 + k_\perp}}$$
where $\theta$ is the angle between the state of $k$ and the axis of the applied field, which in this case is the $z$-axis.

The effective mass, which is the rate of change of momentum w.r.t rate of change of velocity, is $dv_z/dk_z$. Thus we obtain
$$\frac{dv_z}{dk_z} = \frac{dv(k)}{dk_z}\cos\theta-v(k)\sin\theta\frac{d\theta}{dk_z}=\frac{dv}{dk}\cos^2\theta+\frac{v(k)}{k}\sin^2\theta$$

The average over polar angles $\theta$ is
$$<\sin\theta>_\theta=\frac{\int_0^\pi\sin^3\theta d\theta}{\int_0^\pi\sin\theta d\theta}=2/3$$
$$<\cos\theta>_\theta=\frac{\int_0^\pi\cos^2\theta\sin\theta d\theta}{\int_0^\pi\sin\theta d\theta}=1/3$$

Therefore, averaging and taking into account that the field is harmonic we have
$$\delta v_z = -\frac{e}{i\omega\hbar}\left[ \frac{2}{3}\frac{v(k)}{k} +\frac{1}{3}\frac{dv(k)}{dk} \right]F=-\frac{e}{i\omega m_{t}(k)}F$$
where $m_{t}$ is the optical effective momentum dependent mass given by
$$\frac{1}{m_{t}}=\frac{1}{\hbar}\left[\frac{2}{3}\frac{v(k)}{k} +\frac{1}{3}\frac{dv(k)}{dk}\right] = \frac{1}{3\hbar k^2}\frac{d}{dk}(k^2v(k))$$

Notice that the above quantity is the weighted average of the more conventional effective mass $m_*=\hbar^2(\partial^2 E/\partial k^2)^{-1}$ and the transport effective mass $m_*=\hbar^2k(\partial E/\partial k)^{-1}$

Ref: https://www.researchgate.net/publication/340374095_Absorptive_loss_and_band_non-parabolicity_as_a_physical_origin_of_large_nonlinearity_in_epsilon-near-zero_materials

> ## Kane's K.P. Model

From Kane's KP model we have
$$E(k) = \frac{E_{gap}}{2}\sqrt{1+\frac{2\hbar^2k^2}{m^*E_{gap}}}=\frac{E_{gap}}{2}\sqrt{1+(k/k_0)^2}$$
where $m^*$ is the effective mass at $k=0$, $E_{gap}$ is the bandgap energy and $k_0$ is the "non-parabolocity" wave vector given by
$$k_0 = \sqrt{\frac{m^*E_{gap}}{2\hbar^2}}$$
Note that the disperesion relation changes from "parabolic like" to "linear like" at $k=k_0$. 


```python
def E_non_parabolic(c):
  return np.sqrt(1+c**2)-1

plt.figure(figsize=(10, 8))
cs = np.linspace(-2, 2, 1000)
plt.plot(cs, E_non_parabolic(cs), label='Non Parabolic Energy')
plt.plot(cs, 0.5*cs**2, '--', label='Parabolic Approximation')
K = np.array([-1, 1])
plt.plot(K, E_non_parabolic(K), 'o', label='$|k|=k_0$')
plt.title('Energy vs $k_0$', fontsize=20)
plt.ylim(0, 1.5)
plt.xlabel('Wave vector ($k$ in units of $k_0$)', fontsize=15)
plt.ylabel('Energy (in units of $E_{gap}/2$)', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=15, loc='upper right')
plt.show()
```


Using the above relations we can determine the velocity
$$v(k) = \frac{1}{\hbar}\frac{\partial E}{\partial k} = v_{sat}\frac{k/k_0}{\sqrt{1+(k/k_0)^2}}$$
where the saturation velocity is 
$$v_{sat} = \frac{1}{2} \frac{E_{gap}}{\hbar k_0} = \frac{\hbar k_0}{m^*}$$

The momentum dependent optical mass can be calculated from the previously determined expression as:
$$m_t^{-1}=\frac{v_{sat}}{\hbar k_0}\frac{2(k/k_0)^2+3}{3[1+(k/k_0)^2]^{3/2}} = {m^*}^{-1}\frac{2(k/k_0)^2+3}{3[1+(k/k_0)^2]^{3/2}}$$

```python
def velocity_non_parabolic(c):
  return c/np.sqrt(1+c**2)

plt.figure(figsize=(10, 8))
cs = np.linspace(-3, 3, 1000)
plt.plot(cs, velocity_non_parabolic(cs), label='non parabolic band')
plt.title('Velocity vs $k$', fontsize=20)
plt.plot(cs, cs, '--', label='parabolic approximation')
plt.ylim(-2, 2)
plt.xlabel('Wave vector (in units of $k_0$)', fontsize=15)
plt.ylabel('Velocity (in units of $v_{sat}$)', fontsize=15)
plt.legend(fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
def mt_inv(cs):
  return (2*cs**2+3)/(3*(1+cs**2)**1.5)
def mc_inv(cs):
  return 1/(1+cs**2)**1.5

cs = np.linspace(-3, 3, 1000)
plt.figure(figsize=(10,8))
plt.title('$m_t^{-1}, m_{opt}^{-1}, m_c^{-1}$ vs $k/k_0$', fontsize=20)
plt.plot(cs, mt_inv(cs), label='$m_t^{-1}$')
plt.plot(cs, velocity_non_parabolic(cs)/cs,'--', label='$m_{opt}^{-1}$')
plt.plot(cs, mc_inv(cs), '--', label='$m_{c}^{-1}$')
plt.xlabel(r'Wave vector in $k_0$ units', fontsize=15)
plt.ylabel('Inverse Effective Mass', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=15)
plt.show()
```

## Fast Non-linearities in ENZ
Our aim is to find an expression for the polarization induced in the ENZ sample, by the applied electric field $\vec{E} = E_{1}e^{i\omega_{1} t} +  E_{2}e^{i\omega_{2} t}$.
In the Drude-Lorenz model we did this by modelling the electron's motion as a charged simple harmonic oscillator and then by relating the positions amplitudes $x(\omega)$ with induced dipole moments $-e x(\omega)$ and thus tothe susceptibility $\chi(\omega)$.

Here, for a semi-conductor like band structure, in the absence of $\vec{E}$ we have velocity $v$ as a function of wave vector $k$, obtained from the dispersion relation as :

 $$v(k)= \frac{∂E(k)}{\hbar ∂k}$$  

($E$ is energy corresponding to wavevector $k$, not to be confused by electric fields $E_{1,2}$)

Now, assuming the applied fields has perturbative effects on the velocity and wave vectors, in the same we saw in the section on carrier motion, we can expand using Taylor's series in $k$ as:
$$k(t)= k_{0}(t) + δk(t)$$


$$v(k) = \sum_{n=0}^{∞}\frac{∂^{n}v}{∂k^{n}} ⃒_{k_{0}} (k - k_{0})^{n} $$

$$v(k) = v(k_{0}) + \frac{∂v}{∂k} ⃒ _{k_{0}} (\partial k(t))+ \frac{1}{2} \frac{∂^{2} v}{\partial k^{2}}(∂k^{2}(t))+ \frac{1}{6} \frac{∂^{3} v}{\partial k^{3}}(∂k^{3}(t)) + ...$$








We define the fermi-energy as the energy of the highest occupied k-state and the corresponding k-vector as $k_{F}$. 
So the average velocity will be: 

$$<v> = \frac{\Sigma_{k=0}^{k_{F}}v(k)n(k)}{\Sigma_{k=0}^{k_{F(k)}}n(k)}$$ 

where $n(k)$ refers to the multiplicity of that k state

Depending on the convention, we can run the summation from $-k_{F}$ to $k_{F}$ and must modify $n(k)$ appropriately. We note that this doesn't change the final results.




**Average Velcoity**

Now, how would we want to comment on the average velocity or current density? 

(We argue on lines similar to the Drude-Lorenz model, though a different route based on current-density leads to the same results)

We would expect some sort of a weighted sum , the weight corresponding to the multiplicity of the $k$ state.
But now, since we are considering fermions, each k state can be occupied by only one electron.



*Note*: 
We need an approximate for $\chi$ hence it suffices to not use the Fermi-Dirac distribution, asb instead use a uniform distribution of states.Whether we consider spins and say that each state is doubly occupied is not really relevant, as it simply introduces a factor of 2 which we shall see, cancels out in the averages 

**Averages in k-space**

For large $k_{F}$, we can replace the discrete summation by intergration,
$<v> = \frac{\int_{0}^{k_{F}}v(k)n(k) k^{2} dk}{\int_{0}^{k_{F(k)}}n(k) k^{2}dk}$

Here n(k) has been appropriately modified depending on the dimensions of the integrand

Now, recall that n(k) = 1 (or 2, if accounting for spins) ∀ k $ 𝛜$  $ [0,k_{F}]$

Hence, 

$$\int_{0}^{k_{F(k)}}n(k) k^{2}dk = \frac{k_{F}^{3}}{3}$$



Now, we would be substituting v(k) into the above expression, and performing the integration.
Let's see if we can comment on them without evaluating the integrals:



**The average of $v(k_{0})$ term:**

Due to time translational symmtery, $v(-k)=-v(k)$, a sum over all k (or integration over all k) will effectively cancel the corresponding terms, hence we can ignore the first term on R.H.S for the time being.

Now, we recall the calcuated average momentum dependent inverse effetcive transport mass, $\alpha $ and $\beta$, hence we can rewrite,
$$<v> =\hbar[<m^{-1}_{t}> \partial k(t) + \frac{1}{2} <\alpha> ∂k^{2}(t) + \frac{1}{6}<β> \partial^{3}k (t) ] $$


Note: $∂ k (t)$ is independent of k , and is unaffected by the average

**The $ <\alpha> \partial^{2} k$ term:**

From the form of $\alpha$  
we see that $\alpha(k) $ is an odd function. Hence average over k of this term yields 0.


So, we are left with
$<v> =\hbar[<m^{-1}_{t}> \partial k(t) + \frac{1}{6}<β>\partial^{3}k(t)  ] $

**The $<m_{t}^{-1}>$ term :**


From the equation of $m_{t}^{-1}$ in section 3,
$$m_{t}^{-1} =  m^{-1}_{*} \frac{2 (\frac{k}{k_{0}})^{2}+ 3}{3(1 + (\frac{k}{k_{0} })^{2})^{3/2}} $$


$$<m_{t}^{-1}> = \frac{\int_{0}^{k_{F(k)}}m_{t}^{-1}(k) k^{2}dk}{\int_{0}^{k_{F(k)}} k^{2}dk}$$

$$\boxed{<m_{t}^{-1}>=   \frac{m_{*}^{-1}}{(1 + (\frac{k_{F}}{k_{0} })^{2})^{1/2}}  }$$ 


**The  $<\beta>$ term:**

Using the expression for $\beta$:
$$\beta = \frac{m^{-1}_{*}}{k_{0}^{2}} \frac{4(\frac{k}{k_{0}})^4 + 14(\frac{k}{k_{0}})^2 -5 }{3(1+ (\frac{k}{k_{0}})^2)^{7/2}}$$


$$<\beta>= 
 \frac{\int_{0}^{k_{F(k)}}\beta(k) k^{2}dk}{\int_{0}^{k_{F(k)}} k^{2}dk}$$

 $$  \boxed{<\beta>= \frac{m^{-1}_{*}}{k_{0}^{2}}[ \frac{4 \sinh^{-1}(\frac{k_{F}}{k_{0}})}{(\frac{k_{F}}{k_{0}})^3}- \frac{4(\frac{k}{k_{0}})^4 + 11(\frac{k}{k_{0}})^2 +4 }{(\frac{k_{F}}{k_{0}})^2(1+ (\frac{k}{k_{0}})^2)^{5/2}} ]}$$

>## Slow Non-linearities 

The second type of non-linearities that arise is because of the excitation of the electrons to higher levels in the same band. The excitation of these electrons (termed as *hot electrons*) changes the effective mass of the electrons in the fermi level of the conduction band.



Let's see what happens to an electron as it is excited resulting to an energy change $\delta E$.
The momentum $\delta k $ changes, which in turn changes the effective mass $<m_{t}^{-1}>$. This is turn changes the permitivity of the medium, since the relative permittivity of a ENZ material, due to these nearly free carriers, is given by
$$\boxed{\epsilon_{r} = ϵ_{∞} - \frac{\omega_{p}^2}{\omega^{2}+i\omega\gamma} }$$

Where, 

$\omega_{p} = \sqrt{\frac{N e^{2}}{<m_{t}^{-1}>ϵ_{0}(\omega^{2}+ i \gamma \omega )}}$ , is the plasma frequency

$ϵ-{∞}$ is the permittivity at highly non-resonant frequency

$\omega$ is the drive frequency

$N$ is the carrier chare density in the conduction band

**The Power absorbed**

Now, to arrive at change in $<m_{t}^{-1}>$ from the absorbed energy, lets first find the average current density, given as $$J(t) = N e <v(t)>$$
Where, we have an expression of $<v(t)>$ in terms of $\delta k(t)$



From electrodynamics we know that, the absorbed power is given by

$$\boxed{\delta P(t) = E(t) J(t)}$$

We are interested in the $\omega_{1}-\omega_{1}$ component of $P(t)$

So, upto first orders in $J^{(1)}(t)$ and $E(t)$,
we have,
we have $J^{(1)}(t) = N e<v(t)^{(1)}> = N(Ne\hbar <m_{t}^{-1}>\delta k(t)) $

Now, the $\omega_{1,2}$ components of $J^{(1)}$ are:
$$\boxed{J^{(1)}(\omega_{1,2})= i \frac{e^{2}N^{2}}{\omega_{1,2}+ i \gamma_{1,2}} E_{1,2}e^{-i \omega_{1,2}t}  + c.c }$$

So, the total $J$ upto first order in $\omega_{1,2}$ is
$$J^{(1)} = J^{(1)}(\omega_{1}) + J^{(1)}(\omega_{2}) $$

And we have the electric field to be 
$$E = E_{1}e^{i\omega_{1}t} + E_{2}e^{i\omega_{2}t} + c.c$$

So,
we get


$$\boxed{\delta P(t) = \frac{2Ne^{2}\gamma <m_{t}^{-1}>}{\bar\omega^{2}+\gamma^{2}}[E_{1}^{2} + E_{2}^{2} + 2 E_{1}E_{2}cos((\omega_{1}-\omega_{2})t)]}$$

Note: this is the $\omega_{1}-\omega_{2}$ component of the absorbed power, we have other componets arising because of $J^{(3)}$ as well



**The Energy available to hot electrons**

Now, the electrons excited to higher levels (in the same band) relax because of electron-lattice interactions, with a characteristic time $\tau_{el}$

If $U_{hot}$ is the total energy available to these *hot* electrons, we can have the time evolution of $U_{hot}$ as: 



$$\frac{dU_{hot}(t)}{dt} = \delta P(t) - \frac{U_{hot}}{\tau_{el}} $$


Where the net increase in the available energy density increases with the absorbed power $P(t)$ and diminishes due to electron-lattice relaxations, whose characteristic time scale is $\tau_{el}$

$\delta P(t)$ acts as the periodic driving term, and the steady state solution of such a system will also have the same drive frequency


We are interested in soultions of $U_{hot}(t)$ that have the component $e^{-i(2\omega_{1}-\omega_{2})t}$ (because we shall see in a while how it determines the change in effective mass)

This was the reason, we where only interested in the $\omega_{1} - \omega_{2}$ component of $\delta P (t)$, because, if we want a solution for this component of periodic drive, we substitute the expression for $\delta P(t)$ obtained above and have teh solution:
 

$$\boxed{U_{hot}(t) = \frac{\tau_{el}}{1-iΔω\tau_{el}}\frac{2Ne^{2}\gamma <m_{t}^{-1}>}{\bar \omega^{2} + \gamma^{2}} E_{1}E_{2}e^{-i(\omega_{1}-\omega_{2})t} + c.c
}$$



Where $\bar \omega = \omega_{1} + \omega_{2}$ and $Δ\omega = \omega_{1}-\omega_{2}$

Now, this energy is not equally distributed among all the $N$ electrons in the highest level of the conduction band,because hot electrons include both the *primary* electrons generated by the absorption of photons and the *secondary* electrons generated because of the fast electron-electron collisions.



**Relating $U_{hot}$ with change in $<m^{-1}_{t}>$**

Let's say the fraction of hot *primary* electrons  here is $f_{hot}$

So the energy $U_{hot}$ is distributed among $f_{hot} N $ electrons.
The energy of available to a primary electron then is  :
$$\delta E(t) = \frac{U_{hot}(t)}{f_{hot}N} $$

Note: $\delta E(t)$ is not to be confused with electric field

Now, recall from earlier sections that the fermi -velocity of the electron is given by:

$$v_{F} =\frac{1}{\hbar} \frac{∂E}{\partial k} ⃒ _{k=k_{F}}$$

so,$$\delta k(t) = \frac{\delta E (t)}{\hbar v_{F}}$$

Now, this is the change caused in $k$, and we have an expression for $m^{-1}_{t}$ in terms of $\delta_{k}$


$$\delta m^{-1}_{t} =\frac{∂m^{-1}_{t}}{∂k}⃒_{k=k_{F}} \delta k $$

Using the expression for $\delta k$ in terms of $U_{hot}$ ,and recalling our definition of $\alpha$




$$\delta m_{t}^{-1}(t) =  \frac{\partial m_{t}^{-1}}{\partial k}\delta k(t)$$
$$= \frac{\alpha_{F} U_{hot}(t)}{f_{hot}N \hbar v_{F}}$$


**The Non-linear permittivity**


From the Drude's model for premittivity we discussed at the beginning of this section,
$$ϵ_{r}= \epsilon_{\infty} - \frac{ f_{hot} N e^{2} }{ϵ_{0} (\omega^{2} + i \gamma \omega)} <m_{t}^{-1}>$$

a small change in $ϵ_{r}$ will be :

$$\delta ϵ(t)=- \frac{ f_{hot} N e^{2} }{ϵ_{0} (\omega^{2} + i \gamma \omega)} \delta( <m_{t}^{-1}>)$$


Now, 
$$\delta ϵ(t)= \frac{ f_{hot} N e^{2} }{ϵ_{0} (\omega^{2} + i \gamma \omega)} \frac{<m_{t}^{-1}>}{<m_{t}^{-1}>}\delta <m_{t}^{-1}>$$

Now, from the section on fast non-linearities, we know that ,
$$\chi^{(1)}(\omega) = \frac{ f_{hot} N e^{2}<m_{t}^{-1}>}{ϵ_{0} \omega(\omega+ i \gamma)} $$ 

Substituting the expression for $\delta <m^{-1}_{t}>$, 
$$\boxed{\delta \epsilon_{r}(t) ≈ \frac{\alpha_{F} U_{hot}(t) \chi^{(1)}(\bar \omega)}{<m_{t}^{-1}>N \hbar v_{F}} }$$








$<m^{-1}_{t}>$ is averaged over the electronic distribution. As long as the temperatures are not very high, i.e. $k_{b} T_{e} << E_{F}\approx 15,000K$, it suffices not to find the exact electronic distribution, 

We substitute $U_{hot}(t)$. (We had found the $e^{i (\omega_{1}-\omega_{2})t} $ component of $U_{hot}(t)$) Now in the fourier space, the corresponding component of $ϵ(t)$ will be $ϵ(ω_{1}-ω_{2})$, given by,


$$\boxed{\delta ϵ(\omega_{1}-\omega_{2})≈ \alpha_{F}\frac{1}{N \hbar v_{F}}\frac{  \gamma\tau_{el}}{(1-iΔ\omega\tau_{el})} \frac{2N e^{2}}{(\bar\omega^{2}+ \gamma^{2})}\chi^{(1)}E_{1}E_{2}e^{-i(\omega_{1}-\omega_{2})t}}$$







**Commenting on $\chi^{(3)}$**

NNow, recall the discussion on $\chi^{(3)} $.

An incident field $E_{1}e^{i\omega_{1}t}$ experiences the $\delta \epsilon(\omega_{1}-\omega_{2})$ component of permitivitty , giving rise to the non-linear contribution of polarizability $\delta \epsilon(\omega_{1}-\omega_{2})E_{1}e^{-i\omega_{1}t}$ i.e.,  $$ P_{NL}(2\omega_{1}-\omega_{2})= \alpha_{F}\frac{1}{N \hbar v_{F}}\frac{  \gamma\tau_{el}}{(1-iΔ\omega\tau_{el})} \frac{2N e^{2}}{(\bar\omega^{2}+ \gamma^{2})}\chi^{(1)}E_{1}E_{2}e^{-i(\omega_{1}-\omega_{2})t}E_{1}e^{-i\omega_{1}t}+ c.c$$


Now, this is the $e^{i(2\omega_{1}-\omega_{2})t}$ component of non-linear polarizability, which we had defined to be 

$$P_{NL}(2\omega_{1}-\omega_{2}) = \epsilon_{0}\chi(2\omega_{1}-\omega_{2})E_{1}^{2}E_{2}e^{-i(2\omega_{1}-\omega_{2})t}+ c.c$$



Hence, we have ,

$$\chi^{(3)}_{slow}(2\omega_{1}-\omega_{2}) = \frac{2 \alpha_{F}}{ \hbar v_{F}}\frac{  \gamma\tau_{el}}{(1-iΔ\omega\tau_{el})} \frac{ e^{2}}{(\bar\omega^{2}+ \gamma^{2})}\chi^{(1)}(\bar \omega)$$

Let  $\gamma = \tau_{s}^{-1}$

$$\boxed{\chi^{(3)}_{slow}(2\omega_{1}-\omega_{2}) = \frac{2 \alpha_{F}}{ \hbar v_{F}}\frac{  (\tau_{el}/\tau_{s})}{(1-iΔ\omega\tau_{el})} \frac{ e^{2}}{(\bar\omega^{2}+ \gamma^{2})}\chi^{(1)}(\bar \omega)}$$

Note: $\gamma$ accounted for damping in the Drude's Model. $\tau_{s}$ is then called the momentum scattering rate , that accounts for scattering with the phonons

**Finding $\frac{\alpha_{F}}{v_{F}}$**

With the expressions for $\alpha(k)$ and $v(k)$, we have

$$v_{F} = v(k_{F}) = v_{sat} \frac{(\frac{k_{F}}{k_{0}})}{ (1 + (\frac{k_{F}}{k_{0}})^{2})^{\frac{1}{2}} } $$

$$\alpha_{F}=\alpha(k_F)=\frac{{m^{-1}_{*}}}{k_0}(\frac{k_F}{k_0})\frac{2(k_F/k_0)^2+5}{3[1+(k_F/k_0)^2]^{5/2}}$$


$$\boxed{\frac{\alpha_{F}}{v_{F}}= \frac{m^{-1}_{*}}{v_{sat}k_{0} }f_{\alpha}(k_{F})} $$

}$$
​


Now, we put $\frac{\alpha_{F}}{v_{F}}$ in the expression for $\chi^{(3)}$, and have 

$$\boxed{\boxed{\chi^{(3)}_{slow}(2\omega_{1}-\omega_{2};\omega_{1};\omega_{2}) = \frac{ 2 (\tau_{el}/\tau_s)}{ (1-iΔ\omega\tau_{el})} \frac{ e^{2}\chi^{(1)} f_{\alpha}(k_{F})}{\hbar^{2}k_{0}^{2}(\bar\omega^{2}+ \gamma^{2})}}}$$

**Comparing $\tau_{s}$ and $\tau_{el}$**


$\tau_{el}$ , the electron lattice scattering time is $\sim 100 fs$, whereas $\tau_{s}$ is of few  $ 10fs$ or less. We can explain this as follows:

1. $\tau_{s}$ is independent of whether a phonon is absorbed or emitted, it simply accounts for scattering, whereas  $\tau_{el}$ is the time it takes for excited electrons to release energy via phonons(i.e. by interactions with lattice). Hence $\tau_{el}$ depends on the net phonons emitted

2. The energy carried by a phonon is smaller than the thermal energy of the hot electrons, so it takes quite some phonons for electrons to relax

3. The momentum scattering is elastic, unlike lattice relaxation, that is dissipative

## Conclusion 

Our goal in this section is to compare fast and slow non linearities in "enz materials" (semiconductors operating at certain frequencies) to traditional 2 level systems exhitbiting fast and slow non linearities. 
The strength of fast non linearities only depends on the energy difference between the 2 levels (conduction and valence band in this case). This is simiar to traditonal 2 level systems and hence no added advantage is given by enz materials. Intraband non linearities (for enz materials) on the other hand are extremenmly lossy due to the possibility of absorption as for any photon frequency plenty of filled and unfilled states exist. So overall there is no reason to use enz materials over traditional materials for applications required fast non linearities (femtosecond nonlinear
processes, such as supercontinuum generation and optical frequency combs, or for the harmonic generation, the low loss transparent materials are preffered)

The case for slow non linearities is a little different, enz material offer some distinct advantages over traditional materials. Slow non linearities in enz materials are fairly broadband due to the absoption following the smooth drude dispersion. Which is very different from traditional materials which can change orders of magnitude for a few tens of nanometers. 

figure of merit (FOM)—the product of maximum nonlinearity, propagation length, and the signal bandwidth B is useful to optimize materials. For ENZ materials it is given by $\frac{1}{2 \pi \tau_{el}}$ where $\tau_{el}$ is ENZ response time.
The FOM is rather well defined and constant for a given wavelength, therefore a prudent way to optimization leads to making the bandwidth just sufficient for a given task (i.e make it large enough by making response time shorter). In this sense, the ENZ response time, $100 fs < \tau_{el} < 1 ps$, is almost ideal to allow the nonlinear effects to accumulate over the time $\tau_{el}$ while remaining fast enough for many applications in telecommunications such as all-optical switching, adiabatic frequency shifting, and FWM at a speed of a few THz.
For traditiona material the expression for FOM becomes $\frac{1}{2 \pi T_1}$ where $T_1$ is the recombination time which is about 3 orders of magnitude longer hence unsuitable for many applications.

> # Temporal response of a medium  

We consider the temporal response of a medium to an external field $\vec{E}(\omega)$

The polarization produced in the medium at time $t$ can in general depend on the applied electric field at previous times.



##Linear response

Say $R(t,\tau)$ is the linear response function of the medium. It encaptures the contribution to polarization $P(t)$ in the medium at time $t$, due the the effect of electric field applied at some previous time $\tau$

Hence, the $P(t)$ (polarization at time $t$) is given by :  

$$\begin{equation} 
P(t) = ϵ_{0} \int_{-\infty}^{t} d\tau R(t,\tau) E(\tau)
\end{equation}  $$



**Causality of the response function**

As mentioned,  R(t,τ)E(τ)  is the contribution to  P(t)  by the field applied at previous time  τ , and we integrate upto the current time  t 

From causality principle,  P(t)  can't depend on  E(t′)  if  t′>t .

Hence  R(t,t′)=0   ∀t′>t

**Time Translational symmetry of response function**

Suppose we are looking at the contribution of $E(\tau)$ to $P(t)$, i.e., $R(t,\tau)E(\tau)$. 

Now, if the field was not incident at $\tau$ but at $\tau + Δt$, and we were interested in the polarization at $P(t+\Delta t)$; what would the response function be?

By arguments of time translation symmetry, $R(t,\tau) = R(t+\Delta t, \tau + \Delta t)$

We realise that the response function must only depend on the difference between the times $t-\tau$, i.e., $R(t,\tau) = R(t-\tau)$





Now, using the above arguments, 
$$\begin{equation} 
P(t) = ϵ_{0} \int_{-\infty}^{t} d\tau R(t-\tau) E(\tau)
\end{equation}  $$
A change of variables gives us

$$\begin{equation} 
P(t) = ϵ_{0} \int_{0}^{\infty} d\tau R(\tau) E(t-\tau)
\end{equation}  $$


Now, we go to the Fourier space, and define (upto normalization constants)

$E(t) = \int_{-\infty}^{\infty}d \omega$ $ E(\omega)e^{i\omega t} $

$P(t) = \int_{-\infty}^{\infty}d \omega$ $ P(\omega)e^{i\omega t} $






Substituting in the equation for $P(t)$

$$\begin{equation} 
P(t) = ϵ_{0} \int_{0}^{\infty}d\tau R(\tau) e^{i\omega \tau}\int_{-\infty}^{\infty}d \omega E(\omega)e^{- i\omega t} 
\end{equation}  $$

Now, we define  
$$\chi(\omega;\omega) = \int_{0}^{\infty}d\tau R(\tau) e^{i\omega \tau} $$

$\chi(w;w)$ is the susceptibilty in Fourier space


$$\begin{equation} 
P(t) = ϵ_{0}  \int_{-\infty}^{\infty}d \omega  \chi(\omega;\omega) E(\omega)e^{- i\omega t} 
\end{equation}  $$
*Aside*:

And since we have $P(t) = \int_{-\infty}^{\infty}d \omega$ $ P(\omega)e^{i\omega t} $, we see arrive at

$P(\omega)=\chi(\omega;\omega)E(\omega)$

**How fast does the medium respond?**

The arguemnet above lead us to an important equation,
$$\chi(\omega;\omega) = \int_{0}^{\infty}d\tau R(\tau) e^{i\omega \tau} $$

$\chi$ in general depends on the frequency $\omega$ and the response function of the medium at all times 0 to $\infty$  

Now, suppose the medium has a very short response time, such that its instantaneous polarization depends only on the the instantaneous electric field

Such a response function can be modelled by $R(t-t') = c$ $ δ(t-t')$, i.e. it is sharply peaked at $t$, with some weight $c$

Now, the susceptibility $\chi$ of such a medium will be 
$$\chi(\omega;\omega) =c \int_{0}^{\infty} d\tau \delta(\tau) e^{i\omega \tau} = c $$

Thus if the response of the medium is **instantaneous** or **fast**, $\chi$ is **independent of the drive frequency**

If the response of the medium is not **instantaneous** or **slow**, **$\chi$ will depend on $\omega$**


##Non-linear Response function
We can extend the above discussion to non-linear susceptibilities and have,

$$\chi(\omega_{1}+\omega_{2};\omega_{1},\omega_{2}) = ϵ\int_{0}^{\infty}d\tau_{1}\int_{0}^{\infty} d\tau_{2}R(\tau_{1},\tau_{2}) e^{i(\omega_{1} \tau_{1} + \omega_{2}\tau_{2})} $$

Now, if $R(\tau_{1},\tau_{2})=c \delta(\tau_{1})\delta(\tau_{2})$, 

that is , if the response of the medium depends only on the instataneous $E_{1,2}$


$$\chi(\omega_{1}+\omega_{2};\omega_{1},\omega_{2}) = ϵ\int_{0}^{\infty}d\tau_{1}\int_{0}^{\infty} d\tau_{2}R(\tau_{1},\tau_{2})=\int_{0}^{\infty}c \delta(\tau_{1})\delta(\tau_{2}) e^{i(\omega_{1} \tau_{1} + \omega_{2}\tau_{2})} =c $$




### Slow non-linearities
We see that if $\chi$ is a function of $\omega$, the response is **slow** as it depends on the previous states of the system
Such processes physically correspond to the ones associated with *real* excitation of the electrons to higher energy levels. The excited electrons return to ground states after an interlevel relaxation time $T_{1}$

### Fast non-linearities

If $\chi$ is independent of $\omega$ , the response is **fast**, as it depends on instantaneous values of the externally applied field. Such response corresponds to excitation to *virtual* energy levels which are short-lived
