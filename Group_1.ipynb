# Introduction

<p style='text-align:justify;'>    
A number of non-linear effects can be realised with one or more atoms coupled to one or more resonator modes. To achieve this, we need to construct an interaction between an electromagnetic mode and a two-level atom, which is given by the quantum Rabi Hamiltonian. This Hamiltonian conmprises of individual terms for the resonator modes and atoms, and an interaction term that can change the number of excitations in the system, enabling higher-order processes via virtual photons.

In non-linear optics, we require light of high intensity for the non-linear effects to be significant. Similarly, here we require a very strong light-matter coupling for the effects to be noticable. Specifically, the light-matter coupling must be strong enough to ensure that the effective deterministic coupling between system states, induced by the higher-order processes, becomes larger than the relevant decoherence rates in the system.

In this model, we use excitations of resonators modes as opposed to propagating waves used conventionally. This does not require any external drive, and can realise the non-linear effects with a minimum number of photons and potentially unit conversion efficiency.

# Quantum Rabi Model
<p style='text-align:justify;'>    
(Note : Natural units are used throughout this report, i.e. $\hbar = 1, c=1$)

To demonstrate well known non linear optical phenomena like sum frequency generation, second harmonic generation etcetera, we want to describe the simplest possible light matter interaction at the quantum level. The leading candidate for this is the Quantum Rabi Model. The Hamiltonian describing the Quantum Rabi Model is split into 3 terms, the first being  

1. $\hat{H}_f = \Sigma_k\omega _k \hat{a^\dagger}_k \hat{a}_k$

This is obtained by considering a quantised scalar field of photons obeying the Klein-Gordon Equation. For the readers not familiar with Quantum Field Theory, suffice to say that we can show that any quantised field of photons is a superposition of an infinite number of harmonic oscillators of all possible frequencies. Then, for photons of a particular frequency $\omega = ck$, this term simply multiplies the number operator for that mode  $\hat{N} = \hat{a} \hat{a}^\dagger$, with the energy of each photon in that mode. Then we sum over all possible modes.


2. $\hat{H}_A = \frac{1}{2}\omega_q \hat\sigma_z$ 

This describes the energy of the atom independently of the photons. Since by definition of the Rabi model we are considering a 2-state system, we can draw an isomorphism between the 2-state atom and a spin-1/2 particle. Hence, consider a basis $\begin{pmatrix} 0 \\ 1\end{pmatrix}$ (the ground state $g$) and $\begin{pmatrix} 1 \\ 0\end{pmatrix}$ ( the excited state $e$). In this basis, $\hat{\sigma}_z = |e\rangle \langle e| - |g\rangle \langle g|$. Hence, 

$$
\hat{H}_A |e\rangle = \frac{\omega_q}{2} |e\rangle 
$$

$$
\hat{H}_A |g\rangle = -\frac{\omega_q}{2} |g\rangle
$$

Hence we define $\omega_q$ as the transition frequency between the 2 states.

3) $\hat{H}_{int}^{Rabi} = \Sigma_k g(\hat{a}_k + \hat{a}_k^\dagger)(\hat{\sigma}_- + \hat{\sigma}_+)$

This describes the interation between the 2 systems. 

The proof is quite complicated, but it can be summarised as follows:

$$
\hat{H}_{int} = -\hat{\textbf{d}}\cdot\hat{\textbf{E}}
$$
We put in the quantized Electric Field Operator $\hat{\textbf{E}}(\textbf{R})$ as:

\begin{gather}
E(r,t) = i\Sigma_k \Sigma_{\mu = \pm 1} \omega(e^{\mu}(k) \hat{a}_k(t) e^{ik.r} - h.c.)
\end{gather}

We also substitute in the quantised dipole moment operator in terms of the transition dipole moments as follows:
$$
\vec{\hat d} = \hat\sigma_+\langle e|\vec{\hat d}|g\rangle + \hat\sigma_-\langle g|\vec{\hat d}|e\rangle
$$

$$
= \hat\sigma_+\vec d_{eg} + \hat\sigma_-\vec d_{ge}
$$

On simplification, the interaction term reduces to:

$$
\begin{gather} 
\hat{H}_{int}^{Rabi} = \Sigma_kg(\hat a_k + \hat a_k^\dagger)(\hat{\sigma}_- + \hat{\sigma}_+) = \Sigma_kg(\hat a_k + \hat a_k^\dagger)\hat{\sigma}_x
\end{gather}
$$
Where $g$ is a constant determined by the quantised dipole operator $\hat{d}$ of the atom.

Going forward, we drop the indice indicating the mode since there is only one mode.

Now, we introduce the Jaynes Cumming model, which is a simplification of the Quantum Rabi model. We first introduce a new concept, a new picture of quantum mechanics called the Dirac picture (or the interaction picture), which is a intermediate between the Schrödinger and Heisenberg pictures. Here, both wavefunctions and operators change with time.

We split the Schrodinger hamiltonian into two parts, $\hat{H} = \hat{H}_0 + \hat{H}_I$, where the first part is time-independent, and the second part is not. For our current system, $\hat{H}_0$ is simply $\hat{H}_f + \hat{H}_A$. 

Define:

$$
U(t,t_0) = e^{-i\int dt\hat{H}}\\
U_o(t,t_0) = e^{-i\hat{H}_0 (t-t_0)}\\
U_I(t,t_0) = e^{-i\int dt\hat H_I}\\
$$

In the Schrödinger picture:

$$
\begin{gather}
|\psi_S(t)\rangle = \hat{U}(t, t_0)|\psi_S(0)\rangle
\end{gather}
$$

We define the Dirac picture wavefunction as 


$$
\begin{gather}
|\psi_I(t)\rangle = \hat{U}_0^\dagger(t, t_0)|\psi_S(t)\rangle
\end{gather}
$$

i.e. we are removing the evolution due to the time independent part.

To preserve expectation values between pictures, operators are then forced to evolve as:

$$
\begin{gather}
\hat{A}_I(t) = \hat{U}_0^\dagger(t, t_0)\hat{A}_S(t) \hat{U}_0(t, t_0)
\end{gather}
$$

Plug this operator evolution in (1), with $\hat{H}_0 = \hat{H}_f + \hat{H}_A$. We get in the Dirac picture:
\begin{gather}
\hat{H}_{int}(t)=e^{i \hat{H}_0 t / \hbar} \hat{H}_{int} e^{-i \hat{H}_0 t / \hbar}=\hbar\left(g_c \hat{\sigma}_{+} \hat{a}_c^{\dagger} e^{i\left(\omega_c+\omega_{q}\right) t}+g_c^* \hat{\sigma}_{-} \hat{a}_c e^{-i\left(\omega_c+\omega_{q}\right) t}-g_c^* \hat{\sigma}_{-} \hat{a}_c^{\dagger} e^{-i\left(\omega_{q}-\omega_c\right) t}-g_c \hat{\sigma}_{+} \hat{a}_c e^{i\left(\omega_{q}-\omega_c\right) t}\right)
\end{gather}


Now, we assume that $\omega_q \approx \omega_a$. This means that 2 of the terms are fast rotating, while the other 2 are slow. We neglect the rapidly oscillating terms (rotating wave approximation). Going back to the Schrodinger picture, we get:

$$
\begin{gather}
\hat{H}_{int}^{JC} = g(\hat a \hat{\sigma}_+ + \hat{a}^\dagger\hat{\sigma}_-)
\end{gather}
$$

Hence, finally we have:
$$
\begin{gather}
\hat{H}^{Rabi} = \Sigma_k\omega _k \hat{a}_k \hat{a}_k^\dagger + \frac{1}{2}\omega_q \hat{\sigma}_z + \Sigma_k g(\hat{a}_k + \hat{a}_k^\dagger)(\hat{\sigma}_- + \hat{\sigma}_+)
\end{gather}
$$

$$
\begin{gather}
\hat{H}^{JC} = \omega _a \hat{a} \hat{a}^\dagger + \frac{1}{2}\omega_q \hat{\sigma}_z + g(\hat a \hat{\sigma}_+ + \hat{a}^\dagger\hat{\sigma}_-)
\end{gather}
$$
If we have both longitudinal and transverse coupling, the Quantum Rabi Model can be generalised to:
$$
\begin{gather} 
\hat{H}_{int}^{gen} = \Sigma_kg(\hat{a}_k + \hat{a}_k^\dagger)(\hat{\sigma}_x \cos{\theta} + \hat{\sigma}_z\sin{\theta})
\end{gather}
$$

We note that the number of excitations in the Quantum Rabi model always changes by an even number, in the Jaynes-Cummings Model the number of excitations is conserved, and we have no restrictions on the number of excitations in the General Quantum Rabi Model.

Each of these models can be extended to include more resonators or qubits. 



# Vacuum Rabi Oscillations

<p style='text-align:justify;'> 
In this section we demonstrate the effect of strength of coupling between atom and resonator mode on the excitations and frequency of oscillations. This is not a non-linear process yet, since there is only one resonator mode.

```python
!pip install qutip

import numpy as np
import matplotlib.pyplot as plt
from qutip import *
import math
```

We are using the Jaynes Cummings Model for a single radiation mode and two state atom with the rotating wave approximation and no dissipation ,  $ \\ \hat H = \hbar \omega a^\dagger a + \frac \Omega 2 \sigma_z + g\hbar (a \sigma_++a^\dagger \sigma_-)   $,
where $\omega$ is the frequency of the mode, $\Omega$ corresponds to the frequency difference between the two states of the atom and g is interaction strength \
Here, we use units where $\hbar = 1$.

```python
# defining constants (let )
w = 0.2*np.pi #mode frequency
wa = 0.1*np.pi #atom frequency/2

g1 = 0.1
g2 = 0.8
g3 = 4


N = 10 #Number of fock states
t = np.linspace(0, 5, 1000)
```

```python
# defining operators and states
p_0 = tensor(basis(N,0), basis(2,1)) #initial state
a = tensor(destroy(N), qeye(2)) #destruction operator of field
s = tensor(qeye(N), destroy(2)) #destroction operator for atom
```

```python
#setting up Hamiltonians
H1 = w*a.dag()*a + wa*tensor(qeye(N), sigmaz()) + g1*(a*s.dag() + a.dag()*s)
H2 = w*a.dag()*a + wa*tensor(qeye(N), sigmaz()) + g2*(a*s.dag() + a.dag()*s)
H3 = w*a.dag()*a + wa*tensor(qeye(N), sigmaz()) + g3*(a*s.dag() + a.dag()*s)
```

```python
output1 = mesolve(H1, p_0, t, None, [a.dag() * a, s.dag() * s]) 
output2 = mesolve(H2, p_0, t, None, [a.dag() * a, s.dag() * s]) 
output3 = mesolve(H3, p_0, t, None, [a.dag() * a, s.dag() * s])
```

$a^\dagger a$ and $s^\dagger s$ give the excitation probabilities for the resonant mode and atom respectively

```python
plt.figure(figsize = (20,10))
plt.plot(t, output1.expect[0], color = 'blue', label = "cavity, g = 0.1")
plt.plot(t, output2.expect[0], color = 'green', label = "cavity, g = 0.8")
plt.plot(t, output3.expect[0], color = 'pink', label = "cavity, g = 4")
plt.plot(t, output1.expect[1], color = 'blue', linestyle = 'dashed', label = "atom, g = 0.1")
plt.plot(t, output2.expect[1], color = 'green', linestyle = 'dashed', label = "atom, g = 0.8")
plt.plot(t, output3.expect[1], color = 'pink', linestyle = 'dashed', label = "atom, g = 4")
#plt.plot(t, output4.expect[1], color = 'blue', label = "atom")
plt.legend()
str = "Rabi oscillations for different couplings"
plt.title(str)
plt.xlabel('Time')
plt.ylabel('Expectation value of energy')
```

# Effective hamiltonian and $g_\text{eff}$ 

<p style='text-align:justify;'>    
We decompose $\hat H_\text{int}$ into $g_\text{eff}(i,f) |f\rangle \langle i|$ type terms, with the end goal being to maximise a particular $g_\text{eff}(i,f)$ and write $\hat H_\text{int}^\text{eff} = g_\text{eff}(i,f) |f\rangle \langle i| + g^*_\text{eff}(i,f) |i\rangle \langle f|$.

Assuming an $n$-step path is the lowest-order path which can take $|i\rangle$ to $|f\rangle$, to lowest order $g_\text{eff}$ is:
$$
g_\text{eff} = \sum_{j_1,\dots,j_{n-1}}\frac{V_{fj_{n-1}}\dots V_{j_1i}}{(E_i-E_{j_{n-1}})\dots(E_{i}-E_{j_1})}
$$

Up to a phase factor. Here, $V_{ij} := \langle i|\hat H_{int}|j⟩$.

We derive this expression in Appendix A using perturbation theory.

The summation over ${j_1,\dots,j_{n-1}}$ can be intuitively understood as a summation over various paths that can take $|i\rangle$ to $|f\rangle$. A path is a valid one if all pairs of consecutive states in the path $|j_{\mu-1}\rangle,|j_\mu\rangle$ are connected by the interaction hamiltonian - all $V_{j_\mu j_{\mu-1}}$ must be non-zero. 

Now we have an expression for $g_\text{eff}$, which allows us to write the effective interaction hamiltonian. Later, while discussing three-wave mixing, we calculate one of the components (based on one path) for $g_\text{eff}$ and present the complete expression as well.

With knowledge of the expressions of $g_\text{eff}$, we can set up a resonance between $|i\rangle$ and $|f\rangle$ such that all the other lower-order processes reaching different states $|j\rangle$ are far off resonance, and achieve close to unit efficiency for our desired process.

# Three-Wave Mixing

<p style='text-align:justify;'>    
The most general case of three wave mixing in NL optics is sum or difference frequency generation. If we take the quantum picture of SFG and DFG, then in the Fock representation, the creation (sum frequency generation) of a photon with frequency $\omega_+$ can be described as the transition:

$$|{n_1, n_2, n_+}⟩ \rightarrow |{n_1 - 1, n_2 - 1, n_+ + 1}⟩$$

For DFG, the +1's are replaced by -1's.
The interaction Hamiltonian for this sum-frequency generation can be given by (Note that this hamiltonian is for canonical SFG, and not its analogous processes):

$$\hat{H}_{int}^{(+)} = g \hat{a}_1 \hat{a}_2 \hat{a}_{+}^{\dagger} + g^* \hat{a}_1^{\dagger} \hat{a}_2^{\dagger} \hat{a}_{+}$$


Multiple resonator-qubit setups are possible that can serve as analogs to three wave mixing. For example, if we couple three resonators and one qubit, then the transition $|{1, 1, 0, g}\rangle \rightarrow |{0, 0, 1, g} \rangle$ corresponds to SFG, while the transition in reverse corresponds to DFG.



To find the coupling strength of the interaction, we need to compute $\langle{0, 0, 1, g}| \hat{H}_{\text{int}}|{1, 1, 0, g}⟩$. Since the Interaction Hamiltonian is composed of terms with multiple raising and lowering operators, the ket vector will be transformed accordingly. We can conceptualize this process by considering these intermediate steps as _virtual transitions_.



For example, one possible path to SFG can be:

$$|{1, 1, 0, g}⟩ \xrightarrow{\hat{a}_2 \hat{\sigma}_+}
|{1, 0, 0, e}⟩ \xrightarrow{{\hat{a}_+^{\dagger} \hat{\sigma}_-}} |{1, 0, 1, g}⟩ \xrightarrow{\hat{a}_1 \hat{\sigma}_z} |{0, 0, 1, g}⟩$$


We can derive that these virtual transitions combine to give the effective Hamiltonian:


$$\hat{H}^{\text{eff}}_{\text{int}} = g_{\text{eff}}|{0,0,1,g}\rangle \langle{1,1,0,g}| + \text{H.c.}$$


An important benefit of using the an atom coupled to three resonators instead of simply three resonators (as implemented in canonical NL optics) is that we obtain an additional degree of freedom to detune the lower order interactions (caused via $\chi^{(1)}$ instead of the $\chi^{(2)}$ term which is responsible for three wave mixing) to be off-resonance, resulting in the near-unit efficiency of the desired higher-order process.

For example, when the resonance condition for SFG i.e. $\omega_a + \omega_b \approx \omega_c$ is met, we can ensure that the intermediate state $|{1,0,0,e}\rangle$ with the energy $\omega_a + \omega_q$ (which is reachable via a lower order process) will be far off resonance if we choose $\omega_q$ sufficiently far from $\omega_b$.




## Analogous Processes to second harmonic and sub-harmonic generation


Let us take the example of second harmonic and sub-harmonic generation to demonstrate analogous processes.

We can write three such processes:

1.  _Two resonators, One qubit_: $|{1, 0, g}⟩ \rightarrow |{0, 2, g}⟩$
2.  _One resonator, One qubit_: $|{0, e}⟩ \rightarrow |{2, g}⟩$
3.  _One resonator, Two qubits_: $|{1, g, g}⟩ \rightarrow |{0, e, e}⟩$. 


The effective transition for the first process is:
$$\hat{H}^{\text{eff}}_{\text{int}} = g_{\text{eff}}|{0,2,g}\rangle ⟨{1,0,g}| + \text{H.c.}$$

This transition can be achieved via multiple paths, as shown below:

We calculate the component of $g_\text{eff}$ due to one of the paths.

Take the path
$$
|1,0,g\rangle\xrightarrow{\hat b^\dagger\hat\sigma_z}|1,1,g\rangle\xrightarrow{\hat a \hat\sigma_+}|0,1,e\rangle\xrightarrow{\hat b^\dagger\hat\sigma_+}|0,2,g\rangle
$$

Name $i=|1,0,g\rangle, n = |1,1,g\rangle, m=|0,1,e\rangle, f = |0,2,g\rangle $

Then the component of $g_\text{eff}$ is
$$
\frac{V_{fm}V_{mn}V_{ni}}{(E_m-E_i)(E_n-E_i)}
\\
\text{Using the generalised Rabi hamiltonian }\hat H_0 = \omega_a \hat a^\dagger\hat a+\omega_b\hat b^\dagger\hat b + \frac 1 2 \omega_q\hat\sigma_z
\\
\text{And }\hat H_{int} = (g_a(\hat a+\hat a^\dagger)+g_b(\hat b+\hat b^\dagger)(\sigma_z\sin\theta+(\sigma_++\sigma_-)\cos\theta))
\\
V_{fm} = \langle f|\hat H_{int}| m\rangle = \sqrt{2}g_b\cos\theta
\\
V_{mn} = \langle m|\hat H_{int}| n\rangle = g_a\cos\theta
\\
V_{ni} = \langle n|\hat H_{int}| i\rangle = g_a\sin\theta
\\
E_m =  \langle m|\hat H_{0}| m\rangle = \frac 1 2\omega_q+\omega_b
\\
E_n = \omega_a + \omega_b - \frac 1 2 \omega_q
\\ 
E_i = \omega_a - \frac 1 2 \omega_q
\\[0.5ex]
\therefore g_\text{eff}^\text{chosen path} = \frac{\sqrt{2}g_a g_b^2 \cos^2\theta\sin\theta}{(\omega_q+\omega_b-\omega_a)\omega_b}
$$

The complete $g_\text{eff}$, combining the effects of all the paths, is complicated, but it signifies considerably under the resonance condition $\omega_a = 2\omega_b$ to 
$$
g_\text{eff} = \frac{3\sqrt{2}g_ag_b^2\omega_q^2\sin(2\theta)\cos\theta}{4\omega_b^4-5\omega_b^2\omega_q^2+\omega_q^4}
$$

```python
omega_b = 1
omega_q = 1.6*omega_b
g_a = 0.07*omega_b
g_b = 2*g_a
theta = math.pi/6
cavity_fock_states = 3 #don't need more than 3 for our plot
```

```python
a = tensor(destroy(cavity_fock_states), qeye(cavity_fock_states), qeye(2)) #qeye is just identity matrix
b = tensor(qeye(cavity_fock_states), destroy(cavity_fock_states), qeye(2))
sigma_minus = tensor(qeye(cavity_fock_states), qeye(cavity_fock_states), destroy(2)) 

sigma_plus = sigma_minus.dag() #dag() is dagger (adjoint) operator
sigma_x = (sigma_plus + sigma_minus)
sigma_y = 1j*(sigma_minus - sigma_plus)
sigma_z = -0.5j*(sigma_x*sigma_y - sigma_y*sigma_x)
```

```python
def get_eigenenergies(omega_a, choice_str="general"):
    H_0 = omega_a * a.dag() * a + omega_b * b.dag() * b + 0.5*omega_q * sigma_z
    H_int_general = (g_a*(a + a.dag()) + g_b*(b + b.dag()))*(sigma_x*math.cos(theta) + sigma_z*math.sin(theta))
    H_int_rabi = (g_a*(a + a.dag()) + g_b*(b + b.dag()))*sigma_x
    H_int_jc = g_a*(a*sigma_plus + a.dag()*sigma_minus) + g_b*(b*sigma_plus + b.dag()*sigma_minus)
    
    choice_str_to_H_int = {"general":H_int_general, "rabi":H_int_rabi, "jc":H_int_jc}
    H = H_0 + choice_str_to_H_int[choice_str]
    return H.eigenenergies()
```

```python
choices = ("general", "rabi", "jc")
omega_a_values = np.linspace(1.4, 2.1, 400)
eigenenergy_arr = np.array([
    [get_eigenenergies(omega_a, choice_str) for omega_a in omega_a_values]
    for choice_str in choices])
```

```python
choice_to_linestyle = {"general":"solid", "rabi":"dashed", "jc":"dashdot"}
eigenstate_index_to_color = {1:"black", 2:"green", 3:"blue", 4:"red"}
eigenstate_indices = eigenstate_index_to_color.keys() #eigenstate indices for 'H' corresponding to eigenstates |0, 1, g>, |1, 0, g>, |0, 0, e>, |0, 2, g> of 'H_0' 
plt.figure(figsize=(10,10))
plt.ylim(0, 1.3)
plt.xlabel(r'$\omega_a / \omega_b$', size=24)
plt.ylabel(r'$E / \hbar \omega_b$', size=24)

plt.text(1.5, 0.2, '|0, 1, g⟩', size=20)
plt.text(1.5, 0.6, '|1, 0, g⟩', size=20)
plt.text(1.5, 0.9, '|0, 0, e⟩', size=20)
plt.text(1.5, 1.2, '|0, 2, g⟩', size=20)

plt.title("Energy levels for two resonator modes coupled to a qubit", size=15)

def draw_plot():
    for choice_index, choice in enumerate(choices):
        for eigenstate_index in eigenstate_indices:
            #now all that's left is to beautify and plot the lines

            label = None

            if choice_to_linestyle[choice] == "dashdot" and eigenstate_index_to_color[eigenstate_index] == "black":
                label = "J.C. Model"
            elif choice_to_linestyle[choice] == "dashed" and eigenstate_index_to_color[eigenstate_index] == "black":
                label = "Rabi Model"
            elif choice_to_linestyle[choice] == "solid" and eigenstate_index_to_color[eigenstate_index] == "black":
                label = "Generalized Rabi Model"
            line = plt.plot(omega_a_values, eigenenergy_arr[choice_index, :, eigenstate_index]/omega_b,
            linestyle=choice_to_linestyle[choice],
            color=eigenstate_index_to_color[eigenstate_index],
            label=label
            )


draw_plot()

plt.legend(fontsize=15)
```

# Second Harmonic Generation - Plots

We are using the generalised Rabi model for two radiation modes and two state atom with the rotating wave approximation (RWA) and no dissipation. The generalise Rabi model does not impose any restrictions on number of excitations.

The Hamiltonian for the generalized Rabi model is given by:


$ \\ \hat H =  \omega_a a^\dagger a + \omega_b b^\dagger b + \frac \Omega 2 \sigma_z + \hat H_\text{int} \\  $ 

Where $\omega_a$ and $\omega_b$ correspond to the frequencies of the resonator modes, $\Omega$ corresponds to the atom frequency and g is interaction strength. \
We use units where $\hbar = 1$.

$  \hat H_\text{int}  $ for generalized Rabi model:

$ \\ \hat H_\text{int} = [g_a( a + a^{\dagger} ) + g_b(b + b^{\dagger})] \times [\hat \sigma_x \cos(\theta) + \hat \sigma_z \sin( \theta)] \\ $

```python
# defining constants (let )
wa = 0.2*np.pi # 1st resonator mode frequency 
wb = 2*wa # 2nd resonator mode frequency
w = 0.2*np.pi #atom frequency/2
ga = 5
gb = 4
N = 4 #Number of fock states in 1st resonator
M = 4 #Number of fock states in 2nd resonator
t = np.linspace(0, 100, 10000)
```

```python
# defining operators and states
p_0 = tensor(basis(N,2),basis(M,1), basis(2,0)) #initial state
a = tensor(destroy(N),qeye(M), qeye(2)) #destruction operator for resonator 1
b = tensor(qeye(N), destroy(M), qeye(2) ) #destruction operator for resonator 2
s = tensor(qeye(N),qeye(M), destroy(2)) #destroction operator for atom
sx = tensor(qeye(N), qeye(M), sigmax()) 
sz = tensor(qeye(N), qeye(M), sigmaz())
th = 40 #aparametrizes the amount of longitudinal and transversal couplings
```

```python
#setting up Hamiltonian
from cmath import cos, sin

y = (ga*(a+a.dag())+gb*(b+b.dag()))*(sx*cos(th) + sz*sin(th))
H_int = y + y.dag()
H = wa*a.dag()*a +wb*b.dag()*b + w*tensor(qeye(N),qeye(M), sigmaz()) +y

```

## Raman Scattering

<p style='text-align:justify;'>    
Raman Scattering is the inelastic scattering of photons by matter which is a special case of non-degenerate three-wave mixing. Typically, this involves a change in energy of vibrational modes of a molecule due to incident photons from a laser.

There are two types of Raman scattering:\
**Stokes-Raman scattering:**\
The laser photons are shifted to a lower energy state while the vibrational modes gain energy. 
 
**Anti-Stokes scattering:**\
The scattered photon has a higher energy than the incident photons, and the vibrational modes lose energy.

The Raman-scattering process as described above takes place spontaneously; ie; in random time intervals, one of the many incoming photons is scattered by the material. This process is thus called *spontaneous Raman scattering.*

On the other hand, *stimulated Raman scattering* can take place when some Stokes photons have previously been generated by spontaneous Raman scattering, or when deliberately injecting Stokes photons ("signal light") together with the original light ("pump light"). In that case, the total Raman-scattering rate is increased beyond that of spontaneous Raman scattering: pump photons are converted more rapidly into additional Stokes photons. The more Stokes photons that are already present, the faster more of them are added. Effectively, this *amplifies*  the Stokes light in the presence of the pump light.
We consider the following fields: a driving laser (L) mode
of frequency $ω_L$, a Stokes (S) mode of frequency $ω_S$, an
anti-Stokes (A) mode of frequency $ω_A$, and optical vibrational
phonon (V) modes of frequencies $ω_{Vj}$

Raman scattering can be realised through the following interaction Hamiltonians:

1. Stokes Raman scattering

$$
H_\text{int}^\text{(S)} = \sum_j g_{Sj} \hat a_L \hat a_S^\dagger \hat a_{Vj}^\dagger + H.c.
$$

2. Anti Stokes Raman scattering
$$ H_\text{int}^\text{(A)} = \sum_j g_{Aj}^* \hat a_L \hat a_A^\dagger \hat a_{Vj} + H.c. $$

The processes for Four Wave Mixing are extremely similar to that of 3 Wave mixing. We consider 3 types of processes:

1) Type 1: 2 incoming and 2 outgoing signals \
2) Type 2: 3 incoming and 1 outgoing signal \
3) Type 3: 1 incoming and 3 outgoing signals \\

Analogous to 3 wave mixing, the Hamiltonian for optical type 1 mixing is simply:

$$
H_{int} = g(\hat{a}_1\hat{a}_2\hat{a}^\dagger _3\hat{a}^\dagger _4) + g^*(\hat{a}^\dagger _1\hat{a}^\dagger _2\hat{a}_3\hat{a}_4)
$$

and for types 2 and 3:

$$
H_{int} = g(\hat{a}_1\hat{a}_2\hat{a}_3\hat{a}^\dagger _4) + g^*(\hat{a}^\dagger _1\hat{a}^\dagger _2\hat{a}_3^\dagger \hat{a}_4)
$$


### 1) Sum Frequency Generation 
**Analogous Processes**

The process is:

$$
\omega_a + \omega_b \approx \omega_c+ \omega_d
$$

Based on the number of qubits we use to store excitations, we can have different processes for types 1, 2 and 3. For 1 qubit coupled to 4 photons we have:

$$
|1,1,0,0,g\rangle \ \longrightarrow |0,0,1,1,g\rangle
\\
|1,1,1,0,g\rangle \ \longrightarrow |0,0,0,1,g\rangle
\\
|0,0,0,1,g\rangle \ \longrightarrow |1,1,1,0,g\rangle
$$

for type 1, 2 and 3 processes and respectively.
If we store of of the photon excitations in a qubit:

$$
|1,1,0,g\rangle \ \longrightarrow |0,0,1,e\rangle
\\
|1,1,1,g\rangle \ \longrightarrow |0,0,0,e\rangle
\\
|0,0,0,e\rangle \ \longrightarrow |1,1,1,g\rangle
$$

and so on for more qubits. We note that if we store atleast one excitation in a qubit, the number of intermediate steps in each of these processes is reduced (4 for no qubits storing excitations, and 3 for other cases). Hence, it is preferable to use atleast one qubit to store excitations.

### 2) Third Harmonic Generation
We calculate $g_\text{eff}$ for the special case of $3\omega_a = 3\omega_b = 3\omega_c = \omega_d$. Hence for third harmonic generation the transition is $|n, n_+\rangle \longrightarrow |n-3, n_+ + 1\rangle $ and the reverse process for third sub harmonic generation. The interaction Hamiltonian for both of these is:

$$
\hat{H}_{int} = g\hat{a}^3\hat{a}_+ ^\dagger + g^*\hat{a}_+ ^{\dagger^3}
$$

**Analogous Processes** \\
a)	***2 Resonators***: We have 2 photons of frequencies $\omega_a$ and $\omega_b$, where $\omega_a = 3\omega_b$. The required transition then is $|0, 3, g> \leftrightarrow |1, 0, g\rangle$. The interaction Hamiltonian is:

$$
\hat H_\text{int} = (g_a (\hat a+\hat a^\dagger) + g_b(\hat b + \hat b^\dagger) ) \hat \sigma_x
$$

As before, this reduces to an effective interaction Hamiltonian of:
\begin{gather}
\hat{H}_\text{int} = g_\text{eff}|f\rangle\langle i|
\end{gather}

\begin{gather}
g_\text{eff} = \frac{4\sqrt{6}g_ag_b^3\omega_q}{9\omega_b^4 – 10\omega_b^2\omega_q^2 + \omega_q^4}
\end{gather}

b)	***1 Resonator***: the required transition is $|0, e\rangle \leftrightarrow |3, g\rangle$, where $3\omega_a = \omega_q$. Correspondingly the effective interaction Hamiltonian is:
\begin{gather}
\hat{H}^\text{eff}_\text{int} = g_\text{eff}|0 e\rangle\langle 3 g| + h.c.
\end{gather}
Where h.c. stands for hermitian conjugate.


The corresponding $g_\text{eff}$ is obtained as:

\begin{gather}
g_\text{eff} = -\frac{9\sqrt{6}g^3}{4\omega_q^2}
\end{gather}
c)	***3 Identical Qubits***: the required transition is $|0,e ,e, e> \leftrightarrow |1, g, g, g>$, where $\omega_a = 3\omega_q$. The interaction Hamiltonian is:
\begin{gather}
\hat{H}_\text{int} = g(\hat{a} + \hat{a}^\dagger)\Sigma_j \sigma_x^j, j = 1, 2, 3
\end{gather}
Correspondingly the effective interaction Hamiltonian is:
\begin{gather}
\hat{H}^\text{eff}_\text{int} = g_\text{eff}|0, e, e, e\rangle\langle 1, g, g, g| + h.c.
\end{gather}
The corresponding $g_\text{eff}$ is calculated from Third Order Perturbation Theory as:
\begin{gather}
g_\text{eff} = -\frac{3g^3(\omega_a – 3\omega_q}{\omega_q(\omega_a - \omega_q)^2}
\end{gather}
which is 0 for $\omega_a = 3\omega_q$. However, numerically it can be shown that the effective coupling is actually non-zero because of the energy shifting of states induced by the ultra-strong interaction and partly due to the influence of higher-order processes.

# Hyper Raman Scattering

Hyper Raman Scattering is a generalization of Raman scattering to include multiple incoming photons or multiple phonons. 

Hyper Raman scattering based on two photon processes are referred to as type I process. Again this is of two types. For Stokes frequency: 

$$
\hat{H}^{(S)}_{int} = g_{S}\hat{a}_{L}^{2}\hat{a}_{S}^{\dagger}\hat{a}_{V}^{\dagger} + H.c. 
$$

fro anti-Stokes frequency, 

$$
\hat{H}_{int}^{(A)} = g_{A}^{*}\hat{a}_{L}^{2}\hat{a}_{A}^{\dagger}\hat{a}_{V} + H.c.
$$

Analogous processes: 

Here, the qubit excitations play the role of phonons. 

Type-I: two resonators are coupled to a single qubit. ones corresponding to the L mode, one corresponding to the S or A mode  

a) Stokes Hyper-Raman scattering, type I . Setting $\omega_{S} + \omega_{V} \approx \omega_{L}$, the process $|0, 2, g\rangle \rightarrow |1, 0, e\rangle$  corresponds to Stokes hyper-Raman.  

$$
\hat{H}_{int, H_{IS}}^{eff} = g_{eff}|0, 2, g\rangle\langle 1, 0, e| + H.c.$$

$g_{eff}$  for this process is given by 

$$
g_{eff} = \frac{\sqrt{2}g_{a}g_{b}^{2}}{\Delta_{ab}^{2}}
$$

b) Anti-Stokes Hyper -Raman scattering, type I. Setting $\omega_{A} \approx 2\omega_{L} + \omega_{V}$, the transition $|0, 2, e\rangle \rightarrow |1, 0, g\rangle$ corresponds to anti-Stokes hyper-Raman scattering

$$
\hat{H}_{int, H_{IAS}}^{eff} = g_{eff}|0, 2, e\rangle\langle 1, 0, g| + H.c.
$$

For this process, the $g_{eff}$ expression is same as the one obtained for type-I Stokes hyper-Raman scattering despite the resonance condition being different.

# Experimental Feasibility 
Just as non-linear optics usually requires very high light intensity to manifest clearly, the higher-order processes require a very strong coupling to become noticeable. Specifically, the light-matter coupling must be strong enough to ensure that the effective deterministic coupling between system states, induced by the higher-order processes, becomes larger than the relevant decoherence rates in the system.

 In most analogs, the process consists of transferring population from an initial state $|i \rangle $ to a final state $|f ⟩$. The time scale for this process is given by the inverse of the effective coupling strength $g_{eff}$ for the interaction that connects the two states as given by $\hat{H}_{int}^{eff} = g_{eff}|f \rangle \langle i| + H.c.$. For the transfer to be successful, $g_{eff}$ must exceed the relevant decoherence rates in the system:

- Qubit decoherence (relaxation and dephasing) at a rate $\gamma$ and
- Resonator losses at a rate $\kappa$.

i.e.  $g_{eff} > \kappa, \gamma$. 

For it to have close to unit efficiency, $g_{eff} >> \kappa, \gamma$ is required. 

For multiphoton Rabi oscillations, no. of photons must be multiplied to the resonator loss to account the.

Ultrastrong Coupling between light and matter [$g/\omega > 0.1$]has been reached in some solid state experiments. Among these systems, circuit quantum electrodynamics (QED) has provided clearest examples.
In circuit QED with qubits coupled to resonators, their frequencies are usually in the range $\omega_{q}, \omega_{a} \approx  2\pi \ GHz$ . $\gamma$ of the order of $2\pi * 10\ kHz$ have been observed. For some, the Quality factor as given by $Q = \omega_{a}/\kappa$ of order $10^{6}$ can be achieved. 

Essentially, $\gamma, \kappa \approx 10^{-6} \omega_{a}$ can be reached in state-of-the-art circuit QED experiments. Can optimize either the decoherence rate (to be very low) or effective coupling to be very high.

We will work in the Dirac picture (aka interaction picture), where we split the hamiltonian into $\hat{H} = \hat{H}_0 + \hat{H}_I$, the first part being the time-independent non-interaction terms, and the second part being the interaction terms.

Recall that in the Dirac picture, $\hat U_I$ evolves the states, whereas $\hat U_0$ evolves the operators.

$$
\hat U(t,t_0) = e^{-i\int dt\hat{H}}\\
\hat U_0(t,t_0) = e^{-i\hat{H}_0 (t-t_0)}\\
\hat U_I(t,t_0) = e^{-i\int dt\hat{H}_I}\\
$$

So let us expand the state time evolution operator $U_I$ as a [Dyson series](https://en.wikipedia.org/wiki/Dyson_series) (an explanation and derivation is given on the linked page, but intuitively it is similar to a Taylor expansion). Note that $U_I$, being an operator, is also evolved by $U_0$ - we can incorporate that by evolving $H_I$ by $U_0$ and using the obtained $H_I(t)$ in the expressionf or $U_I(t)$.

$$

Our final goal is to calculate $|\langle f|U_I(t,t_0) |i\rangle|^2$, which can be written as : 
$$
|\langle f|\hat U_I(t,t_0) |i\rangle|^2 = |\sum_{n=0}^\infty \langle f| \hat U^{(n)}_I(t,t_0) |i\rangle|^2
$$

Let the terms with $n\lt N$ for some $N$ be zero, which would happen if there does not exist a path of $n$ steps such that each $\langle j_\mu| \hat H_I |j_{\mu-1}\rangle $ is non-zero (i.e. each pair of states $|j_\mu\rangle, |j_{\mu-1}\rangle$ is coupled by the interaction hamiltonian).

The terms with $n\gt N$ have extra factors of the form $g/\Delta E$. We retain the term lowest-order in these factors, which is $n=N$. Then

$$
P(|i\rangle\rightarrow|f\rangle, t) = |\langle f|\hat U_I(t,t_0) |i\rangle|^2 \approx |\langle f| \hat U^{(N)}_I(t,t_0) |i\rangle|^2
\\
= \frac{\left(e^{i(E_f-E_i)t}- 1\right)^2}{(E_{i}-E_{f})^2} \left|\sum_{j_1,\dots,j_{n-1}}
  \frac{V_{fj_{n-1}}\dots V_{j_1i}}{(E_i-E_{j_{n-1}})\dots(E_{i}-E_{j_1})}\right|^2
$$

Is the probability of the transition, which in the limit $t\rightarrow \infty$ gives the transition rate:
$$
W_{(|i\rangle\rightarrow|f\rangle)} = 2\pi\delta(E_f-E_i)\left|\sum_{j_1,\dots,j_{n-1}}
  \frac{V_{fj_{n-1}}\dots V_{j_1i}}{(E_i-E_{j_{n-1}})\dots(E_{i}-E_{j_1})}\right|^2
$$

Which seems to merely imply conservation of energy in the transitions. But this result is more useful than that. 

Recall that we decomposed $\hat H_I $ into $g_\text{eff}(i,f) |f\rangle \langle i|$ type terms.

We apply Fermi's Golden Rule, which states that the transition rate between two states goes as

$$
\tau_{i\rightarrow f} = 2\pi |\langle f|\hat H_I|i\rangle |^2 \delta(E_f-E_i)
$$

Comparing with the transition rate we derived, we obtain

$$
g_\text{eff} = \sum_{j_1,\dots,j_{n-1}}\frac{V_{fj_{n-1}}\dots V_{j_1i}}{(E_i-E_{j_{n-1}})\dots(E_{i}-E_{j_1})}
$$

Up to a phase factor.

# Appendix B - Decoherence rate 
## Qubit Decoherene: 
The atom / qubit won't stay in the state it is prepared infinitely due to coupling with surrounding environment and fluctuations of control parameter in quantum operations. 
It is characterized by the coherence time which is defined as the limit on how long a qubit can retain its quantum state before that state is ruined by noise. It depends on where the qubit lives. We need the quantum state to live longer than the time it takes to perform an operation or experiment. Isolating  the quantum state from surroundings such as enclosing the atoms in ultra high vacuum chambers is one way to do that. 
Coherence rate is inversely proportional to coherence time, long coherence time corresponds to small coherene rate. 

## Circuit-QED
Superconducting circuits with Josephson junctions can behave as artificial atoms. They can be used as qubits. Superconducting qubits over natural atoms because of the control of tunability of qubit parameters. but this also means having to deal with additional coupling channels to the environment, and this unwanted coupling can increase the fragility of nonclassical states. 

Decoherence should be small enough to observe an energy level anticrossing due to effective coupling between states.
