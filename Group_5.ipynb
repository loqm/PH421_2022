# Introduction 

### What does this paper look at? 

The paper analyses triply resonant optical parametric oscialltors (OPO) in semi-conductor microcavities in the strong coupling regime. We have focussed on interpretting the models that describe these systems and understanding the minimum threshold energy required to sustain the oscialltion in the system. Several equations have been plotted to reflect the behaviour of the system and solve equations. 

#### What is a triply resonant system?

In a cavity with 2 mirrors, we send 3 beams - singal, pump and idler. In a singly resonant system, either the signal or idler is reflected by both mirrors (ane effectively gets trapped in the cavity). Similarly, for triply resonant systems, the signal, idler and pump get reflected by the mirrors. 

#### What is strong coupling?

Interaction between two systems being stronger than the individual systems, results in the formation of a quasi particle (photon + exciton = polariton) 



### Setting up the oscialltor system
#### DBR a.k.a. distributed Bragg Reflector 

**The experimental setup described in the paper consisits of a planar distributed Bragg Reflector microcavity, with a few quantum wells.**

It consists of multiple layers of alternating materials with varying refractive index. Each layer boundary causes a partial reflection of an optical wave (read Fresnel reflection - When a light beam  reaches an interface between two different transparent media, it is partly transmitted into the other medium and partly reflected back into the original medium). 
<p align="center" style="max-height:10em;">

</p>
The reflected parts have a phase shift of 180 only if the incident light goes from low-index medium in a high-index medium. The relative phase difference of all reflected beams is zero or a multiple of 360 and therefore they interfere constructively. Therefore, it is used as the equivalent of a mirror in an optical cavity (i.e. the space between two mirrors) to reflect lasers. 

#### Microcavity
The space between two reflectors (i.e. mirrors) is known as a cavity. These cavities are of the order of micrometers (wavelenght of visible light is in the range of 380 nm to 700 nm) to fit less than one whole wavelength, therefore are called microcavities. 

#### Quantum Well
A very thin layer that traps electrons in its plane so that it cannot escape to its surrounding regions. When a light wave of appropriate frequency passes through the quantum well, it can provide enough energy to excite an electron from the atom it is bound to, creating a vacancy. This vacancy is positively charged relative to the surroundings and is called a hole. Since the electron and the hole have opposite charges, the attract and form a new particle known as **exciton**! 

#### Polaritons

Eventually, the electron 'falls into' the hole and the energy is released as a light wave with the same frequency as the incident wave. The micro'resonator' is calibrated in such a way that the light wave and the exciton have the same energy. That light wave converts an electron into an exciton, which collapses to give rise a light wave..... and the cycle repeats. A repeating cycle between light waves and exciton is made possible. This entire system is known as a POLARITON. 
<p align="center">
</p>

Polaritons combine the very strong third-order optical nonlinearity originating from exciton-exciton interactions to the dispersion relation, as a function of the in-plane wave vector k. 

This makes polaritons the perfect candidate for triply-resonsnat OPO applications. 

**This papers establishes a mean field description of the cavity-polariton field equations, in terms of a third-order non-linear wave equation.**

#### Mean Field Theory 

An approximation technique that allows us to map a multi-body problem onto a one-body problem, by replacing the N-body system with a 1-body system that sits in a suitably chosen external field. This external field is set equal to the average field due to the remaining particles.

### We try to understand the dispersion relations of the entire system in order to describe and understand it better, before defining it mathematically

#### Avoided Crossing
We take a quick detour tour here to understand a theoretical concept known as **Avoided Crossing**

It is important to study two-level systems in quantum mechanics, because most physical systems can be demoposed into a two body system. We can observe the effect of perturbation on a two-state system through avoided crossings when you plot the individual energy vs energy difference curve of the eigenstates. The two-state Hamiltonian can be written as $ H = \begin{pmatrix}E_1 & 0 \\ 0 & E_2 \end{pmatrix} $ with eigenvectors $ \begin{pmatrix}1 \\ 0 \end{pmatrix} $ and $ \begin{pmatrix}0 \\ 1 \end{pmatrix} $. These two eigenvectors represent the two states of the system. If $E_1$ is equal to $E_2$, the system is degenerate. If $\psi_1$ and $\psi_2$ are eigenstates of the system, it implies that $\hat{H}\psi_1 = E_1\psi_1$ and $\hat{H}\psi_2 = E_2\psi_2$. If $E_1$ is equal to $E_2$ then $\hat{H}a_1\psi_1 + \hat{H}a_2\psi_2 = E(a_1\psi_1 + a_2\psi_2)$, where $a_1\psi_1 + a_2\psi_2$ is an eingenstate of the Hamiltonian too. Therefore a system prepared in this state, will remain in the same state (with only phase differences being accumulated, which doesn't change the state).

However, when subjected to an external perturbation, the matrix elements of the Hamiltonian change. For the sake of simplicity we consider a perturbation with only off diagonal elements. Since the overall Hamiltonian must be Hermitian we may simply write the new Hamiltonian as $ H' = H + P = \begin{pmatrix}E_1 & 0 \\ 0 & E_2 \end{pmatrix} + \begin{pmatrix}0 & W \\ W* & 0 \end{pmatrix}  = \begin{pmatrix}E_1 & W \\ W* & E_2 \end{pmatrix} $

The modified eigenvalues can be found by diagonalising this matrix, which yields 

$E_+=\frac{1}{2}\left(E_1+E_2\right)+\frac{1}{2}\sqrt{\left(E_1-E_2\right)^2+4|W|^2}$ 

$E_-=\frac{1}{2}\left(E_1+E_2\right)-\frac{1}{2}\sqrt{\left(E_1-E_2\right)^2+4|W|^2}$

#### How to use this to plot the dispersion relation of the system? 

Dispersion relation is the plot between the energy of the system and the wave number. If we consider our polariton to be nothing but a two body system, we can treat it like the above with the laser exciting the electrons in the quantum well as the perturbing force. $E_1$ and $E_2$ will be. Therefore, $E_+$ and $E_-$ represent the upper and lower polariton respectively. Susbsituting the values of $E_1$ and $E_2$ in the above equation yields the realtion between $k$ and $E$. 

#### Hopfield Relation
##### An alternative model to analyse our two body system

We begin with constructing the Hamiltonian for this system. The Hopfield dielectric theory allows us to construct a model of dielectrics which consist of quantum harmonic oscialltors which interact with the modes of the quantum electric fields (i.e. photons in this case). Therefore, the Hamiltonian of the excitons and the optical cavity can be written using ladder operators, similar to that of a quantum oscilalltor's energy. 

$$ \begin{align}
\hat{H}_{C} &= \sum_{\vec{k}} E_c(\vec{k}) \hat{a}^\dagger_{\vec{k}} \hat{a}_{\vec{k}} \\ 
\hat{H}_{X} &= \sum_{\vec{k}} E_x(\vec{k}) \hat{b}^\dagger_{\vec{k}} \hat{b}_{\vec{k}} \\ 
\hat{H}_{\text{coupling}} &= \sum_{\vec{k}} \frac{1}{2} \hbar \Omega_R \left( \hat{a}^\dagger_{\vec{k}} \hat{b}_{\vec{k}} + \hat{b}^\dagger_{\vec{k}} \hat{a}_{\vec{k}} \right)  \\ 
\hat{H}_{\text{total}} &= \hat{H}_{C} + \hat{H}_{X} + \hat{H}_{\text{coupling}} \\ 
&= \sum_{\vec{k}} \left(E_c(\vec{k}) \hat{a}^\dagger_{\vec{k}} \hat{a}_{\vec{k}} + E_x(\vec{k}) \hat{b}^\dagger_{\vec{k}} \hat{b}_{\vec{k}} + \frac{1}{2} \hbar \Omega_R \left( \hat{a}^\dagger_{\vec{k}} \hat{b}_{\vec{k}} + \hat{b}^\dagger_{\vec{k}} \hat{a}_{\vec{k}} \right) \right)
\end{align}$$

where $E_c(k)$ is proportional to $hw_c$ and $E_x(k)$ is proportional to $hw_x$. 

$$
E_{LP}(\vec{k})=\frac{E_{C}(\vec{k})+E_{X}(\vec{k})}{2}+\sqrt{\Omega_R^2+\left(\frac{E_{C}(\vec{k})-E_{X}(\vec{k})}{2}\right)^2} \\
E_{LP}(\vec{k})=\frac{E_{C}(\vec{k})+E_{X}(\vec{k})}{2}-\sqrt{\Omega_R^2+\left(\frac{E_{C}(\vec{k})-E_{X}(\vec{k})}{2}\right)^2}$$

The above are the dispersion relations for the upper polariton and the lower polariton respectively. 

A distinctive observation is Hamiltonian of the coupling term. We can observe an energy exchange between both the systems i.e. $a_k^\dagger$ increments the energy of the cavity photon system and the $b_k$ operator decrements the energy of the exciton system. The inverse happens in the second term. This energy exchange is reflective of the coupling between both the systems. The magnitude of the coupling is mediated by the Rabi coupling energy $\hbar \Omega_R \sim 2.5 \text{meV}$ for this paper.

Now, we need to diagonalise the matrix. We can do this by writing it in terms of new eigenvectors. Hopfield transformation hop to our rescue again (excuses for the pun!). 

```python
@interact(k_p1=(0,10,0.1), w_c1=(0,5,0.1), k_p2=(0,10,0.1), w_c2=(0,5,0.1), W=(0,2,0.1))

def PolaritonenergyPlot(k_p1=1, w_c1=1.5,k_p2=10, w_c2=1.5,W=1):
    x=np.arange(-10,10,0.1)
    #global k_p1,k_p2,W
    #print(E_1)
    
    E_1=w_c1 * np.sqrt((x/k_p1)**2+1)
    E_2=w_c2 * np.sqrt((x/k_p2)**2+1)

    E_UP=(1/2)*(E_1+E_2)+(1/2)*(np.sqrt((E_1-E_2)**2+4*W**2))
    E_LP=(1/2)*(E_1+E_2)-(1/2)*(np.sqrt((E_1-E_2)**2+4*W**2))

    plt.plot(x, E_1 , linestyle='--',color='blue')
    plt.plot(x, E_2 , linestyle='--',color='red')

    plt.plot(x, E_UP , linestyle='-',color='blue')
    plt.plot(x, E_LP , linestyle='-',color='red')
    
    plt.xlim([-10, 10])
    plt.ylim([-0,20])
```

### We then set up the equations that describe our system

##### Assumption

The dynamics is relevant only for the lower polariton branch, implying that the number of photons in the upper polariton branch is neglible. This helps us simplify the discussion by focussing only on the former.

The laser is monochromatic and circularly polarized. The circular polarization is preferred because it is preserved by nonlinear interactions and is almost completely transferred to the signal, idler beams. 

##### Driven Damped Harmonic Oscillator describing this system 

$$ \begin{matrix}i\frac{d}{dt}\psi_{LP}(\mathbf{k})=\left[{\epsilon}(k)-i\frac{\gamma(\mathbf{k})}{2}\right]\psi_{LP}(\mathbf{k})+F_p(\mathbf{k})e^{-i\omega_pt}\ +\sum_{\mathbf{q}_1,\mathbf{q}_2} g_{\mathbf{k},\mathbf{q}_1,\mathbf{q}_2}\psi_{LP}^\ast\left(\mathbf{q}_1+\mathbf{q}_2-\mathbf{k}\right)\psi_{LP}\left(\mathbf{q}_1\right)\psi_{LP}\left(\mathbf{q}_2\right)\end{matrix} $$
With the exciton-exciton coupling constant being taken as momentum independent $\overline{g}$, and with the exciton delocalized over the number of quantum wells present, we obtain, with $g = \frac{\overline{g}}{N_{QW}}$,
$$g_{\mathbf{q_1,q_2,k}} = gX^*(\mathbf{k})X^*(\mathbf{q_1+q_2-k})X(\mathbf{q_1})X(\mathbf{q_2})$$
And taking the number
$$F_p(\mathbf{k}) = _{\mathbf{k,k_p}}C(\textbf{k}_p)\sqrt{\frac{\gamma_{rad}I_{inc}}{N_{tr}\hbar _p}}$$
This equation is no different from a driven damped oscillator. Inituitively, we can understand 
the first term to be the Hamiltonian of the system (where $\epsilon(k)$ is the energy of the system and $-i\frac{\gamma(k)}{2}$ is the damping term)
the second term as the pump (which drives the system), and
the last term is the coupling term which exibits the non linear charcateristics of the field.

Polariton OPOs require lesser energy for excitation than classical OPOs. The key objective of this paper is to determine the minimum amount of energy that must be supplied to the system to sustain the oscillation. 

We begin by considering a pump only solution i.e. among the different $k$ modes, only the mode at $k$ = $k_p$ contains a source term in the equation. This leads us to assume an ansatz of $$\psi_{LP}(x,t) = Pe^{i(k_px - \omega_pt)}$$. Substituting this in the above equation i.e. the driven damped harmonic oscillator describing the system, we get a condition on the amplitude P. 

$$[\epsilon_p - \omega_p - \frac{i}{2}\gamma_p + g|X(k_p)|^4|P|^2]P + F_P = 0$$

Here, $\epsilon_p$ is the frequency of the pump we input; $\gamma_p$ is the corresponding linewidth (which is a material dependent propoerty). 

We have two more useful relations to work with, 

$$ n_{xp} = |X(k_p)|^2|P|^2 $$ and, 
$$ I_P = |F_P|^2 $$

We can use this to establish the relation between $I_P$ (driving intensity) and $\frac{gn_{xp}}{\gamma}$. This relation is plotted in Fig 2. 

```python
def I_p_calc(n_xp, X_p, g, E_p, w_p, y_p) :
  return n_xp / (X_p)**2 * ( (g * X_p**2 * n_xp + E_p - w_p)**2 + (y_p / 2)**2 )

@interact(k_z=(0,5,0.1),k_p=(0,10,0.1),w_c=(0,5,0.1),W=(0,2,0.1),g=(0,5,0.1),y_p=(0,5,0.1),X_p=(0,1,0.01))
def bistabilityfunc(k_z=1.54,k_p=5.3,w_c=1.57,W=0.546,g=1,y_p=1,X_p=0.8):
    #print(P)
    #print(E_LP)
    
    w_p = E_1 =w_c*np.sqrt((k_p/k_z)**2+1)
    E_2=w_c
    
    E_p=(1/2)*(E_1+E_2)-(1/2)*(np.sqrt((E_1-E_2)**2+4*W**2))


    #print(I)
    #plt.xlim([-10, 10])
    #plt.ylim([-0,2])
    if (w_p - E_p > np.sqrt(3)/2 * y_p) :
      print("We have the optical bistability case")

      x = (w_p - E_p) / y_p
      n_xp_lower = ( 2*x - np.sqrt(x*x - 3 / 4) ) / 3 / X_p **2
      n_xp_upper = ( 2*x + np.sqrt(x*x - 3 / 4) ) / 3 / X_p **2

      y_lower  = np.append(np.arange(0, n_xp_lower, 0.1), n_xp_lower)
      y_middle = np.append(np.arange(n_xp_lower, n_xp_upper, 0.1), n_xp_upper)
      y_upper  = np.arange(n_xp_upper, n_xp_upper+2, 0.1)

      x_lower  = I_p_calc(y_lower , X_p, g, E_p, w_p, y_p)
      x_middle = I_p_calc(y_middle, X_p, g, E_p, w_p, y_p)
      x_upper  = I_p_calc(y_upper , X_p, g, E_p, w_p, y_p)

      plt.plot(x_lower , (g*y_lower)/y_p , c='b')
      plt.plot(x_middle, (g*y_middle)/y_p, c='r')
      plt.plot(x_upper , (g*y_upper)/y_p , c='b')
    else :
      print("We have the optical limiter case")

      y_whole = np.arange(0, 10, 0.1)
      x_whole  = I_p_calc(y_whole , X_p, g, E_p, w_p, y_p)
      
      plt.plot(x_whole , (g*y_whole)/y_p , c='b')
```

#### Observations from the plot

In Fig 2(a), $n_xp$ monotnically increases as a funcrion of the driving intensity $I_p$. This occurs when the pump frequency is below or close to the resonance, i.e. 
$$ \omega_p - \epsilon(\mathbf{k}_p) < \frac{\sqrt{3}\gamma(\mathbf{k}_p)}{2} $$

In Fig 2(b), we observe bistability which we attribute to hysteretic bheaviour due to positive non linear feedback. This occurs in the region where $$ \omega_p - \epsilon(\mathbf{k}_p) > \frac{\sqrt{3}\gamma(\mathbf{k}_p)}{2} $$

For increasing laser intensity, the pump mode population follows the lower branch until its endpoint is reached, and then it jumps to the upper branch. 

If the driving intensity is later ramped down, the system keeps following the upper branch until its end point, and only here the pump mode population jumps back to the lower branch. 

Hysteretic behavior is obvious here, as the upward and downward jump points do not coincide.

### Stability of the Pump-Only State 

In nonlinear dynamical systems, finding a solution is not sufficient, as one has to verify its dynamical stability. The instability of the central branch in Fig 2(b) reflects the dynamic unstablity of the system. 

We introduce two perturbations to judge the stability of the system i.e. signal and idler waves.  
Perturbed solution in $k$-space: 
$$ \Psi_{LP} = Pe^{-i\omega_pt}\delta_{k,k_p}+ u(k_s)e^{-iAt}\delta_{k,k_s}+ v^*(k_s)e^{-iBt}\delta_{k,k_i} $$
with $A = \omega_p + \omega(k_s)$, $B = \omega_p - \omega^*(k_s)$, $k_i = 2k_p-k_s$. We choose $u$ and $v$ in such a way that we can interpret them as the signal and idler waves later in the paper. 

As usual, we substitute our ansetz in the equation of the driven damped oscillator describing the system. It is simpler to look at the operators individually and then put it all together.\\
Operating $i\frac{d}{dt}$ gives:

$$ Pe^{-i\omega_pt}\omega_p\delta_{k,k_p}+ u(k_s)e^{-iAt}A\delta_{k,k_s}+ v^*(k_s)e^{-iBt}B\delta_{k,k_i} $$

The summation would have been tricky, if not for the $\delta$-functions we got from our fourier-transform. This helps us pick out the terms from the summation that have non-zero contributions. Let's pick out the triples $(q_1,q_2,k)$ that give non-zero contribution. We have:

$$(k_p,k_p,k_p),\biggl[(k_p,k_s,k_p),(k_s,k_p,k_p)\biggr],\biggl[(k_p,k_i,k_p),(k_i,k_p,k_p)\biggr],(k_p,k_p,k_s),(k_p,k_p,k_i)$$

Where we have bracketed the terms that contribute the same quantity. Finally, linearising gives us an eigen-value equation for the column vector $w(k_s) = [u(k_s),v(k_s)]^T$

\begin{equation}
L(k_s)w(k_s) = \omega(k_s)w(k_s)
\end{equation}

with the matrix:

\begin{pmatrix}
\epsilon_s - \omega_p + 2g|X_i|^2|X_p|^2|P|^2-i\frac{\gamma_s}{2}& gX^*_sX_iX_p^2P^2\\
-gX_sX^*_iX^{*2}_pP^{*2} & -\epsilon_i + \omega_p - 2g|X_i|^2|X_p|^2|P|^2-i\frac{\gamma_i}{2}
\end{pmatrix}

Now we need imaginary part of the eigenvalue to be negative for stability. We know this from the theory of non-linear dynamics. Normally the condition would be applied on the real part of the eigenvalues, but we need to compensate for the factor of $i$ that was present on the LHS. The imaginary part of the eigenvalues are:

$$\text{Im}(\omega_{\pm}(k_s)) = -\frac{\gamma_s + \gamma_i}{4} \pm\text{Im}{\sqrt{\Biggl((\epsilon_{si}-\omega_p) + g(|X_s|^2 + |X_i|^2)n_{xp} - i\frac{\gamma_s-\gamma_i}{4}\Biggr)^2-g^2|X_s|^2|X_i|^2n_{xp}^2}}$$

Notice that the idler and signal frequencies only appear in the form of their average, $\epsilon_{si} = \frac{\epsilon_s+\epsilon_i}{2} $



### Two kinds of physically distinct instabilities can arise. 

#### Single Mode Instability 

A single mode instability arises when the equation of motion for the pump mode alone i.e. $k_s$ = $k_p$ (which implies that $k_i$ = $k_p$ = $k_s$). This instability involves the $k_p$ mode only, and for this reason it is called single-mode. 

If the imaginary part of $\omega_+(k_s)$ is positive, then over time, the perturbation grows high enough to reach the limits of the small perturbation region (if this equation was true outside of this assumption, then it would grow to infinity). 

There are two ways to verify that the pump only solution 
1. The existence of a saddle node bifurcation - Since two fixed point suddenly appear, we know that it is a saddle node type bifurcation. We know that the lower branch forms a stable fixed point as an excitonic population appearing spontaneously near $I_p=0$ would be aphysical. Since the stable lower and upper arms meet the middle arm of the curve at the two saddle node bifurcations, the only logical conclusion is that the middle arm is unstable while the upper branch is stable.

2. Substituting $k_i$ = $k_p$ = $k_s$ into the $\operatorname{Im}(w_+(k_s))$ equation to conclude that it is positive rather than negative 

#### Parametric instability 

This instability occurs when $k_s \neq k_p$. This signals the onset of parametric oscillations with a finite intensity appearing in a pair of distinct signal/idler waves. 

### Available range of signal/idler frequencies 

We limit ourselves to the study of the lower threshold for parametric emission. For each value of pump wave vector $k_p$, it is important to characterize the range of $\epsilon_{si}$ that can be obtained when the signal/idler wave vectors $k_{s,i}$ are spanned through all different polariton states. 

#### Small $k_p$
For small $k_p$, the $\epsilon_{si}$ vs $k_s$ curve has a single minimum at $k_s = k_p$ where $\epsilon_{si} = \epsilon_p$, and then tends to a finite limit for large $k_s$. 

#### Large $k_p$
For larger values of $k_p$, negative values $\epsilon_{si} - \epsilon_p < 0$ can be reached. The minimum is in fact split in two separate minima, symmetrically located around the '_pump angle_' as required by the symmetry of $epsilon_{s,i}$ under exchange of the signal and idler modes. The upper limit of the available band mono- tonically decreases as a function of $k_p$, due to the corresponding increase of $\epsilon_p$. In particular, it tends to 0 for large values of $k_p$.


```python
from scipy import optimize

def E_lp(k_p,k_z,w_c,W):
    E_1=w_c*np.sqrt((k_p/k_z)**2+1)
    #E_11=w_c*sqrt((k_p/k_z)**2+1)
    E_2=w_c
    

    E_UP=(1/2)*(E_1+E_2)+(1/2)*(np.sqrt((E_1-E_2)**2+4*W**2))
    E_LP=(1/2)*(E_1+E_2)-(1/2)*(np.sqrt((E_1-E_2)**2+4*W**2))
    return E_LP

def E_diff(k_p,k_s,k_z,w_c,W) :
  return ( E_lp(k_s,k_z,w_c,W) + E_lp(2 * k_p - k_s,k_z,w_c,W) ) / 2 - E_lp(k_p,k_z,w_c,W)

@interact(k_z=(0,10,0.1),k_p=(0,10,0.1),w_c=(0,10,0.1),W=(0,2,0.1),y_p=(0,5,0.1))
def fig3(w_c=2,k_z=1.57,W=0.82,k_p=0.1,y_p=0.5):
    fig, (plot1, plot2) = plt.subplots(1,2, sharey=True, figsize=(12, 7))
    
    x_1=np.arange(-10,10,0.01)
    y_1 = E_diff(k_p, x_1, k_z,w_c,W) / y_p
    
    plot1.axvline(x=k_p)
    plot1.plot(x_1,y_1,'p')

    x_2=np.arange(-0.5,7,0.01)
    
    y_2 = [scipy.optimize.minimize_scalar(lambda k_s : E_diff(k_p, k_s, k_z,w_c,W), bounds=[-1,k_p+1]) for k_p in x_2]
    y_min = [E_diff(k_p,res.x, k_z, w_c, W) / y_p for res,k_p in zip(y_2,x_2)]
    y_max = (w_c - E_lp(x_2, k_z, w_c, W)) / y_p
    
    #print(y)
    plot2.axvline(x=0,c='k')
    plot2.axhline(y=0,c='k')
    plot2.plot(x_2,y_min,c='m')
    plot2.plot(x_2,y_max,c='m')
    plot2.fill_between(x_2, y_min, y_max, facecolor='m', alpha=0.5)
    
```



#Calculating the pump mode population $n_{xp}$ at the parametric threshold. 

$$\text{Im}(\omega_{\pm}(k_s)) = -\frac{\gamma_s + \gamma_i}{4} \pm\text{Im}{\sqrt{\Biggl((\epsilon_{si}-\omega_p) + g(|X_s|^2 + |X_i|^2)n_{xp} - i\frac{\gamma_s-\gamma_i}{4}\Biggr)^2-g^2|X_s|^2|X_i|^2n_{xp}^2}}$$

We begin by setting $\gamma_s = \gamma_i$ in the above equation. 
This will give as the easiest analysis as the expression under the square root will be purely real.

At the threshold point, the value under the square root should be zero. If it is positive, the system will be unstable and if it is negative the system will be stable. 

$${\Biggl((\epsilon_{si}-\omega_p) + g(|X_s|^2 + |X_i|^2)n_{xp}\Biggr)^2=g^2|X_s|^2|X_i|^2n_{xp}^2}$$

Plotting the above equation we get a graph of $\frac{g n_{xp}}{\gamma}$ vs $\frac{\omega_p - \epsilon_{si}}{\gamma}$ 



```python
@interact(X_s=(0,1,0.01),X_i=(0,1,0.01))
def bistabilityfunc(X_s=1, X_i=1):
  x = np.arange((X_s**4 + X_i**4 + X_s**2 * X_i**2) / ( 4 * X_s**2 * X_i**2), 3, 0.1)

  y =  ( x *(X_s**2 + X_i**2) - np.sqrt((x * X_s * X_i)**2 - (X_s**4 + X_i**4 + X_s**2 * X_i**2)/4) )/ (X_s**4 + X_i**4 + X_s**2 * X_i**2) 
  plt.plot(x,y)
  b,t = plt.ylim()
  plt.ylim(0,t)
  plt.xlim(0,3)
```

By applying the energy conservation condition, $\epsilon_{si} - \epsilon_p < 0$, i.e. input energy must be greater that signal and idler energies, we get the minima of $n_{xp}$ to occur just after the point where Parametric oscillations cannot occur.

By differentiating the expression above and setting the derivative of ${\omega_{p} - \epsilon_{si}}$ to zero, we get the expressions.

$$\omega_p = \epsilon_{si} + g(|X_s|^2+|X_i|^2)n_{xp}$$ 
and
$$n_{xp}^{min} = \frac{\gamma_{si}}{2g|X_s||X_i|}$$

We also see that for the resonance condition $${2\omega_{p} = \omega_{i}+\omega_{s}}$$ with the frequencies of the signal and idler having a detuning given by
$${\omega_{s} = \epsilon_s + 2g|X_i|^2|X_p|^2|P|^2}$$
$${\omega_{i} = \epsilon_i + 2g|X_s|^2|X_p|^2|P|^2}$$

we get back the expression of the minima. Hence, the minima occurs at resonance.

Combining the above equations we get the optimal detuning to be
$$\omega_p - \epsilon_{si} = \gamma_{si}\frac{|X_s|^2+|X_i|^2}{2|X_sX_i|}$$



##  Laser Intensity at threshold and possible behaviours
We observe that with the presence of the two dynamic instabilites, we can obtain a diverse and rich range on behaviours depending on the relative overlap and positioning of the regions featuring the two behaviours.
___
Single-mode instabilities can only occur when the corresponding eigenvalues from the perturbation matrix are have growth term greater than 0, i.e. $$
-\frac{\gamma_p}{2} + \sqrt{g^2|{X_p}|^4 n_{xp}^2- \left(\epsilon_{p} - \omega_p + 2 g |{X_p}|^2 n_{xp}\right)^2 } \geq 0
$$
Here we have taken the positive branch of the square-root since that is the only one that can be positive. This relates to the following range of values for $n_{xp}$, $$ \frac{ 2 \frac{\omega_p - \epsilon_{p}}{\gamma_p} - \sqrt{\left(\frac{\omega_p - \epsilon_{p}}{\gamma_p}\right)^2 - \frac{3}{4}} }{3 X_p^2} \leq \frac{g n_{xp}}{\gamma_p} \leq 
\frac{ 2 \frac{\omega_p - \epsilon_{p}}{\gamma_p} + \sqrt{\left(\frac{\omega_p - \epsilon_{p}}{\gamma_p}\right)^2 - \frac{3}{4}} }{3 X_p^2}$$
The bistability region occurs only if the argument of the square root is positive, i.e. $\omega_p - \epsilon_{p} > \frac{\sqrt{3}}{2} \gamma_p$, ignoring the other solution branch as the current assumption of the exciton being strongly resonant with the photon energies eliminates the possibility of $\omega_p < \epsilon_{p}$. We define the variables such that the condition is equivalent to $$ n^{\text{bi}}_{xp,\text{lower}} \leq n_{xp} \leq n^{\text{bi}}_{xp,\text{upper}} $$
___

For paramteric oscillations, the corresponding eigenvalue equation provides the inequality (asusming $\gamma_{p,s,i} = \gamma$),  $$-\frac{\gamma}{2} + \sqrt{g^2{X_s}^2{X_i} n_{xp}^2- \left(\epsilon_{si} - \omega_p + g ({X_s}^2 + {X_i}^2) n_{xp}\right)^2 } \geq 0$$
This results in the range of allowed $n_{xp}$ as $$ \frac{ 2 \frac{\omega_p - \epsilon_{si}}{\gamma_p} \left({X_{s}}^{2} + {X_{i}}^{2}\right) - \sqrt{\left(\frac{\omega_p - \epsilon_{si}}{\gamma_p}\right)^2 {X_{s}}^{2}{X_{i}}^{2} - \frac{1}{4}\left({X_{s}}^{4} + {X_{i}}^{4} + {X_{s}}^{2}{X_{i}}^{2}\right) } }{\left({X_{s}}^{4} + {X_{i}}^{4} + {X_{s}}^{2}{X_{i}}^{2}\right)} \\
\leq \frac{g n_{xp}}{\gamma_p} \leq \\
\frac{ 2 \frac{\omega_p - \epsilon_{si}}{\gamma_p} \left({X_{s}}^{2} + {X_{i}}^{2}\right) + \sqrt{\left(\frac{\omega_p - \epsilon_{si}}{\gamma_p}\right)^2 {X_{s}}^{2}{X_{i}}^{2} - \frac{1}{4}\left({X_{s}}^{4} + {X_{i}}^{4} + {X_{s}}^{2}{X_{i}}^{2}\right) } }{\left({X_{s}}^{4} + {X_{i}}^{4} + {X_{s}}^{2}{X_{i}}^{2}\right)} $$
with a corresponding condition on when OPO can actually occur. Again, we specify similarly defined variables such that the condition is equivalent to $$ n^{\text{OPO}}_{xp,\text{lower}} \leq n_{xp} \leq n^{\text{OPO}}_{xp,\text{upper}} $$
___
Thus the 5 major regions of operation, from the perspective of prioritizing OPO, can be described as follows :
1. <font color='red'>Region A</font> : Bistability does not occurs but OPO does \\
   Here the exitonic population increases monotonically with increasing pump intensity and thus it acts exactly like the $\chi^2$ materials, except for the fact that there is an upper bound for the intensity where OPO can occur. \\
   This region is marked <font color='red'>red</font> in the graph below. \\
   Region of interest : $\frac{w_p - \epsilon_{si}}{\gamma_p} > \frac{1}{2} \left( {\frac{X_s}{X_i}}^2 + {\frac{X_i}{X_s}}^2 + 1\right), \frac{w_p - \epsilon_{p}}{\gamma_p} < \frac{\sqrt{3}}{2}$

2. <font color='darkseagreen'>Region B</font> : Bistability occurs but OPO happens at larger intensities \\
   Here, bistability in the single-mode solution does occur, but it occurs before the OPO region and without any overlap, thus the two effects are still absolutely decoupled. However, the pump detuning being adequate to introduce instability raises the minimum intensity threshold, making this is a less favorable operating region than A with regards to OPO. \\
   This region is marked <font color='darkseagreen'>green</font> in the graph below. \\
   Region of interest : $\frac{w_p - \epsilon_{si}}{\gamma_p} > \frac{1}{2} \left( {\frac{X_s}{X_i}}^2 + {\frac{X_i}{X_s}}^2 + 1\right), \frac{w_p - \epsilon_{p}}{\gamma_p} > \frac{\sqrt{3}}{2}, n^{\text{bi}}_{xp,\text{upper}} < n^{\text{OPO}}_{xp,\text{lower}}$

3. <font color='mediumpurple'>Region C</font> : Bistability occurs and OPO overlaps with the upper branch \\
   In this region the $n_{xp}$ needed for OPO have receeded enough so that they overlap with the bistability region's upper branch. Any overlap between OPO and single-mode instability region actually results in parametric oscillations because as the amplitudes of $k_s$ and $k_i$ grow, they deplete the pump mode population and pull it off the unstable pump-only solution. This is further explored in Section 4. \\
   Operation in this region is favorable from the perspective that since oscillation starts right after the we reach the lower saddle node bifurcation point, we jump to oscillations, which can be sustained for a wide range of intensities, thus fine-tuning is note required. \\
   This region is marked <font color='mediumpurple'>purple</font> in the graph below. \\
   Region of interest : $\frac{w_p - \epsilon_{si}}{\gamma_p} > \frac{1}{2} \left( {\frac{X_s}{X_i}}^2 + {\frac{X_i}{X_s}}^2 + 1\right), \frac{w_p - \epsilon_{p}}{\gamma_p} > \frac{\sqrt{3}}{2}, n^{\text{bi}}_{xp,\text{lower}} < n^{\text{OPO}}_{xp,\text{lower}} < n^{\text{bi}}_{xp,\text{upper}} < n^{\text{OPO}}_{xp,\text{upper}}$

4. <font color='slategray'>Region D</font> : OPO occurs completely inside the single-mode instability region region \\
   This region is not of particular interest since we cannot reach the OPO region with steady change in the pump intensity, since the hysteresis loops runs circles around the region where OPO can occur. This region is still accessible via discontinuous pumping however. \\
   This region is marked <font color='slategray'>gray</font> in the graph below. \\
   Region of interest : $\frac{w_p - \epsilon_{si}}{\gamma_p} > \frac{1}{2} \left( {\frac{X_s}{X_i}}^2 + {\frac{X_i}{X_s}}^2 + 1\right), \frac{w_p - \epsilon_{p}}{\gamma_p} > \frac{\sqrt{3}}{2}, n^{\text{bi}}_{xp,\text{lower}} < n^{\text{OPO}}_{xp,\text{lower}} < n^{\text{OPO}}_{xp,\text{upper}} < n^{\text{bi}}_{xp,\text{upper}}$

5. <font color='burlywood'>Region E</font> : OPO occurs before the bi-stability region \\
   This region sees the upper ramp of intensity enter parametric oscillations before we actually enter the bistability region. Depending on the exact values this can be an issue, as the OPO region is quite small for reasonable paramter values and very close to the hysteresis turning point, such that small power fluctuations could send us to the stable upper branch, requiring a full cycling of the intensity before we can start OPO again. \\
   This region is marked <font color='burlywood'>orange</font> in the graph below. \\
   Region of interest : $\frac{w_p - \epsilon_{si}}{\gamma_p} > \frac{1}{2} \left( {\frac{X_s}{X_i}}^2 + {\frac{X_i}{X_s}}^2 + 1\right), \frac{w_p - \epsilon_{p}}{\gamma_p} > \frac{\sqrt{3}}{2}, n^{\text{OPO}}_{xp,\text{lower}}  < n^{\text{bi}}_{xp,\text{lower}} < n^{\text{OPO}}_{xp,\text{upper}} < n^{\text{bi}}_{xp,\text{upper}}$

Technically two other regions exist, christened Regions F where the OPO region is entirely before the sing;e-mode instability with no overlap, and H where the single-mode instability is subsumed into the OPO region. However for our interest in OPO, they are effectively identical to Region E and don't merit a proper distinction.

## Lowest possible threshold intensity
In the last section we derived the detunings which change $\epsilon_{si}$ required to minimize the $I_p$ for a given $k_p$. In order to obtain the lowest possible $I_p$ where parametric oscillation starts, we now impose a condition on $k_p$ too, although indirectly. 
From the pump-only solution, we know that $$F_p = - P \left(g \left|X_p\right|^2  n_{xp} + \epsilon_p - \omega_p - i \frac{\gamma_p}{2}\right) $$
Since $\gamma_p$ is a material property, minimizing the magnitude of the multiplier between $F_p$ and $P$ leads to the real part being set to zero, giving us the detuning condition for the pump, and a simplified relation between the $n_{xp}$ and $I_p$, $$\omega_p  = \epsilon_p + g \left|X_p\right|^2  n_{xp} \\
I_p = \frac{\gamma_p^2 n_{xp}}{4 \|{X_p}|^2} $$
Combining, this gives us the values for the detunings derived in the last section for the lowest possible required intensity, $$ \Delta^\text{opt}_p=\omega_p - \epsilon_p = \frac{\gamma}{2} \frac{{|X_p|}^2}{|{X_s X_i}|} \\
\Delta^\text{opt}_{si,p} = \epsilon_{si} - \epsilon_p = -\frac{\gamma}{2} \frac{{X_s}^2+{X_i}^2-{X_p}^2}{{X_s X_i}} \\ 
I_p^\text{min} = \frac{\gamma^3}{8g {X_s X_i} {X_p}^2}$$
Further assuming that radiative decay rate is the as the polaritonic one, $$ I^\text{min}_\text{inc} = \frac{N_{tr}N_\text{QW}}{8 \overline{g}} \frac{\gamma^2 \hbar \omega_p}{{X_s X_i} {X_p C_p}^2}$$
