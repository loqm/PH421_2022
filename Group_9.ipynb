# **Bistable all-optical devices based on nonlinear epsilon-near-zero (ENZ) materials.**

### **Introduction**

Bistability is a property of system that exhibits two stable states. It rests one of those states depending on history of the system. It is the underlying principle for switching logic and storage applications. There were many electronic and magnetic bistable devices but in optical domain there is no device with acceptable characteristics(fast speed,low power,wide optical bandwidth). Most of proposed devices are based on combination of non-linear optical material and a resonant cavity. These require high power for switching and often incompatible with integrated photonics. Lack of reliable bistable optical devices prevent the technological in both conventional and quantum optics.

Now with the use of Epsilon-Near-Zero (ENZ) material properties, we can make up a device which doesn't have a resonator, enabling broadband operation, and operate at sub-ps switching speed.


### **Epsilon-Near-Zero Materials**

As name suggests, these are optical non linear materials which has very low value of dielectric permittivity. This property makes us to manipulate and control the progation of radiation on subwavelength level.

Unique properties for ENZ materials are:

**1.Large refractive index tuning:**

**2.Enhanced harmonic generation**

**3.Near unity generation of phase conjugate waves**

Transparent oxides possess two salient features which makes them different from other non-linar optical materials.

**1.High degree of electric field confinement**

**2.Ideal balance of strength and speed**

For ENZ materials, enchancement of nonlinearity related to dielectric constant as

The change of dielectric constant is given by 
$$Œ¥Œµ ‚àº œá^{(3)}E^{2} $$

where œá(3) is third-order nonlinear susceptibility.

The electric field is related to the intensity of light I as

With these points in mind, we can consider the enhancement of nonlinearity in ENZ having two factors ‚Äì intrinsic (or material) $œá^{(3)}$, and extrinsic $1/n^{2}$.


##Momentum Dependent Mass
The pump energy is absorbed by the free carriers, raising their kinetic energy and the electron temperature $T_{e}$. The average effective mass depends on the occupied state distribution $œÅ(E)^{.}f(E)$. As the distribution of carriers shifts to higher energy states, the average effective mass of the electron sea tends to increase. This in turn, causes a decrease in the plasma frequency and a corresponding increase in dielectric constant. 

Now to derive the momentum dependent mass of the electron, we consider how the momentum of electrons at the Fermi surface change due to an applied electric field F. Here, the field shifts the Fermi-surface in k-space, producing a change in momentum that for some state k is given by: 
$$
Œ¥\textbf{k} = -\frac{e\textbf{F}}{\hbar}
$$
For a field applied along the z-direction, the velocity associated with a particular momentum state becomes:
$$
v_{z} = v_{F}cosŒ∏ = v_{F}\frac{k_{z}}{\sqrt{k^{2}_{z}+ k^{2}_{\perp} }}
$$
where Œ∏ is the angle between state $k$ and the axis of the applied field, in this case, the z-axis. We can then introduce the effective mass of the electron, described by the rate of change of the velocity versus the rate of change of the momentum, $dv/dk$. Thus, when we take the full-derivative with respect to $k_{z}$ we obtain:
$$
\frac{dv_{z}}{dk_{z}} = \frac{dv(k)}{dk}cos^{2}Œ∏ + \frac{v(k)}{k}sin^{2}\theta
$$
Then averaging over the polar angle, and taking into account that the field is harmonic, we obtain:
$$
\delta{v_{z}} = -\frac{e}{i\omega\hbar}\left[\frac{2}{3} \frac{v(k)}{k}+\frac{1}{3} \frac{d v(k)}{d k}\right]F = -\frac{e}{i\omega m^{*}(k)}F
$$
where:
$$m^{*}(k)=\hbar\left[\frac{2}{3} \frac{v(k)}{k}+\frac{1}{3} \frac{d v(k)}{d k}\right]^{-1}
$$
The above expression gives us the momentum dependent mass of the electron.

### **Average Effective Mass of the Electron**


The effective mass of the electron sea can be calculated by averaging the effective mass of each electron $m(E)$ across all the energy states. The distribution is given by the Fermi-Dirac probability $f\left(E, \mu, T_{e}\right)$. 

$$m_{a v g}^{*}=\left[\frac{1}{N V} \sum_{l=1}^{n_{c a r}} \frac{1}{m^{*}\left(k_{l}\right)}\right]^{-1}$$
For continuous energy band, it's given as follows:
$$
{m_{ {avg}}^{*}}=\frac{1}{N} \int \frac{f\left(E, \mu, T_{e}\right)}{m^{*}(E)} \rho(E) d E
$$
The expression for Fermi-Dirac distribution is:
$$
f\left(E, \mu, T_{e}\right)=\left[\exp \left(\frac{E-\mu}{k_{B} T_{e}}\right)+1\right]^{-1}
$$
where $f\left(E, \mu, T_{e}\right)$ is the Fermi-Dirac distribution, N is the carrier concentration, and $\rho(E)$ is the density of states calculated earlier. 
The applied intensity can modify the the chemical potential $\mu$ and the electron temperature $T_{e}$. Hence the appropriate expression for the intensity dependent effective mass is:
$$
{m_{ {avg}}^{*}(I)}=\int \frac{f\left(E, \mu(I), T_{e}(I)\right)  \rho(E)}{m^{*}(E)} d E
$$

##Dependence of Permitivity on Frequency


The optical response of the material under an applied field of arbitrary frequency is calculated by summing the induced
displacements for all electrons in the gas, $z(t) = -i\omega^{-1}v(t)$, as:

$$
\varepsilon(\omega)=\varepsilon_{\infty}-\frac{q^{2}}{\varepsilon_{0}\left(\omega^{2}+i \omega \Gamma\right)} \frac{1}{V} \sum_{l=1}^{n_{c a r}} \frac{1}{m^{*}\left(k_{l}\right)}=\varepsilon_{\infty}-\frac{N q^{2}}{\varepsilon_{0}\left(\omega^{2}+i \omega \Gamma\right) m^{*}_{avg}}
$$

This the generalized Drude formula for non-parabolic bands, where $Œì$ is the scattering rate, l is an occupied state, and $n_{car}$ is the number of carriers in the volume V. The dependence on intensity comes from the change in average effective mass, hence the dependence of effective mass on intensity must be calculated first.

When an external pump is applied, for any given pump intensity, the distribution of carriers described through the Fermi-Dirac
probability can be found using conservation identities of charge and energy. Firstly, the total number of carriers in the band must remain constant 

$$
\int f\left(E, \mu(I), T_{e}(I)\right) \rho(E) d E=N$$


Secondly, the total change in energy of the electron gas must be equal to the absorbed energy from the pump

$$
\int f\left(E, \mu(I), T_{e}(I)\right) E \rho(E) d E-\int f\left(E, \mu_{0}, T_{0}\right) E \rho(E) d E=\delta \mathcal{U}_{a b s}=A I \tau / d
$$
 ;    where $\mu_{0}$ and $T_{0}$ are the chemical potential and the electron temperature before pumping. $\tau$ is given by the electron relaxation time.



The above two conditions must be satisfied for any given intensity. Hence, the chemical potential and electron temperature can be unquely defined for any given intensity. This enables us to calculate the intensity dependent effective mass as discussed earlier:

$$
{m_{ {avg}}^{*}(I)}=\int \frac{f\left(E, \mu(I), T_{e}(I)\right) \rho(E)}{m^{*}(E)} d E
$$

The effective mass has been shown to vary linearly, increasing ~ 20% per $ TW/{cm}^2$ of applied pump intensity

Finally, we can calculate the the permittivity as a function of both $\omega$ and intensity

$$
\varepsilon(\omega, I)=\varepsilon_{\infty}-\frac{N q^{2}}{\varepsilon_{0} m^{*}_{avg}(I)} \frac{1}{\omega^{2}+i \omega \Gamma}
$$

where,
$$\epsilon_{\infty}=\frac{N q^{2}}{\epsilon_{0} m^{*}_{avg}(0)\omega_{p}^2}$$


### **Bistability with ENZ waveguide**



ENZ materials exhibit a dispersion of its relative electrical permitivitty that combines responses of bound ($ Œµ_{‚àû} $) and free electrons according to 
$$ 
Œµ(\omega) = Œµ_{‚àû} - \frac {œâ^{2}_{p}}{\omega^{2} + iœâ\gamma}
$$

where $Œµ_{‚àû}$ is permittivity due to the bound electrons, $\omega_{p}= (Ne^{2}/Œµ_{0} m^{*})^{1/2}$ is the plasma frequency, $N$ is the carrier density, $m^{*}$ is the energy (or momentum) dependent effective mass and $\gamma$ is the scattering rate. 

The real part of
permittivity changes from positive to negative at the so-called ENZ frequency $œâ_{0}=œâ_{p}/Œµ_{‚àû}^{1/2}$ (identified as
screened plasma frequency) with an imaginary part of the permittivity introducing absorption in the
material.

According to boundary conditions for light polarized normal to interface (TM), the displacement *D=ŒµE* is continuous across the interface, and therefore the electric field (and hence the energy density) is much higher in the ENZ material than in the conventional dielectric. If a thin layer of ENZ material is inserted in the slot of a waveguide (which can be photonic or plasmonic), the closer is the wavelength to $œâ_{0}$, the higher is the field the slot. This is a well-studied slot waveguide effect that allows subwavelength field confinement in the low permittivity medium, taken to its ultimate $Œµ$~$0$ limit. 

Unlike the dielectric cladding, ENZ material is highly absorptive and as the optical field concentrated inside the slot increases, the absorption increases. At the point of large electric fields, the nonlinear properties of the TCO-based ENZ materials come into play. When light is absorbed by the TCO material, it causes an increase of the electron temperature in the conduction band, that in turn leads to an increase of the effective mass and corresponding rise of the dielectric constant. If the carrier density is such that in the absence of light the real permittivity is $ Re(Œµ)< 0 $, an increase of $Re(Œµ)$ would move it closer to zero, which in turn will cause even higher concentration of energy inside the ENZ material layer. Thus, a positive feedback mechanism, essential for bistability, is established. As more energy is concentrated in the ENZ layer, the electron temperature in it rises causing an increase of the dielectric constant and further concentration of energy in the ENZ layer. Obviously, no stable solution can exist with positive feedback and switching occurs.

We can provide an estimate for the strength and dependence of the nonlinearity by taking a few assumptions in order to obtain an analytical expression for the nonlinear coefficients in the ENZ region. We assume
that the average effective mass of the electron gas is approximately equal to the effective mass at the Fermi Level,
$ m^{*} _{avg} ‚âà m^{*} (k_{F}) $ , and that the velocity is nearly saturated at the Fermi level, i.e.,

 $$ \frac{dv(k_{F})}{dk} << \frac{v(k_{F})}{ k_{F}} $$

When electrons absorb a pump photon, their energy changes by some amount  $\delta E_{i}$, which for electrons near the Fermi level produces a change of momentum for the electron of $\delta k_{i} = \delta E_{i} / \hbar v_{F}$ , where $v_{F}$ is the Fermi velocity

 Given our momentum dependent effective mass, the change in the effective mass due to the
intraband absorption process is:

$$ 
\delta \left(\frac{1}{m^{*}}\right)_{i} = \left(\frac{‚àÇ}{‚àÇk} \frac {1}{m^{*}(k)}\right)_{i}\delta k = - \left(\frac{1}{m^{*}(k_{F})}\right) \frac{\delta k_{i}}{k_{F}} = -\left(\frac{1}{m^{*}}\right) \frac{\delta E_{i}}{\hbar v_{F}k_{F}}
$$

Although $\delta E_i$, and therefore $\delta k_i$, is different for each electron, the sum of the energy change over all of the electrons within a given volume is equal to $V^{-1}\sum\limits_i\delta E_i=\delta \mathcal{U}_{abs}$, where $\delta \mathcal{U}_{abs}$ is the absorbed energy density of the pump beam. We can then find the total change to the permittivity by summing over the small modifications in the effective mass induced by the excitation of hot electrons which while neglecting loss becomes:

$$\delta ùúÄ \approx -\dfrac{q^2}{ùúÄ_0 \omega^2} \dfrac{1}{V} \sum\limits_{\delta k} \delta \left(\dfrac{1}{m^*}\right)_i=\dfrac{q^2}{ùúÄ_0 m^* (k_F)\omega^2}\dfrac{\delta \mathcal{U}_{abs}}{\hbar v_F k_F}$$

and when using the relation $ùúÄ_{\infty}=Nq^2/(ùúÄ_0 m^*_{avg}\omega^2_p)$ reduces to:

$$\delta ùúÄ(\omega_p) = ùúÄ_{\infty}\dfrac{\delta\mathcal{U}_{abs}}{N\hbar v_F k_F}$$   

at the screened plasma frequency.

Note that $\delta\mathcal{U}_{abs}$ is same as  $\mathcal{U}_{hot}$ which is the energy density transfered to the hot carriers in the ITO. 


Thus using the above equations we get,
$$ \frac{\delta m^{*}}{m^{*}} = \dfrac{\mathcal{U}_{hot}}{N\hbar v_F k_F}$$   


Using the relation between plasma frequency and effective mass, we get

$$
\frac{\delta m^{*}} {m^{*}} = -\frac{\delta œâ^{2}_{p}}{œâ^{2}_{p}} = \dfrac{\mathcal{U}_{hot}}{N\hbar v_F k_F}
$$

For an optical power P in the waveguide, the absorbed power density in the ITO filled slot is $ \mathcal{U}_{hot} = P_{abs} œÑ = \alpha P \tau/ S_{ITO}$, where $\tau$ is the "cooling time", i.e., the time it takes to transfer energy from the hot carriers to the lattice via electron-phonon scattering.

For the sake of simplification, we assume that $ƒßk_{F}\nu_{F} ‚âàE_{F}$ which is physically valid ‚Äì in order to change the effective mass $\mathcal{U}_{hot}$ of each electron in an assembly of $N$ electrons, the change of energy of each electron $\mathcal{U}_{hot}/N$ should be noticeable relative to the energy it already had, i.e., roughly Fermi energy. We then obtain the following linearized relation between the absorption coefficient, plasma frequency and power.
$$
œâ^{2}_{p} = œâ^{2}_{p_0} - œâ^{2}_{p_0} \frac{\alpha \tau}{NE_{F}S} P
$$
Where $œâ_{p0}$ is the plasma frequency in the absence of optical power and heating.

```python
class slab:
    def __init__(self, alpha_max, alpha_min, power_data):
        self.l = 1e-6
        self.dz = 1e-8
        self.alpha = np.zeros(int(self.l/self.dz))
        self.power = np.zeros(int(self.l/self.dz))
        self.alpha_arr_max = np.array(alpha_max)
        self.alpha_arr_min = np.array(alpha_min)
        self.power_data = np.array(power_data)
        self.last_p = 0

    def calc_alpha(self, power, p_up = True):
        idx = np.argmin(np.abs(self.power_data - power))
        try:
            if len(idx) > 1:
                idx = idx[0]
        except:
            pass

        if p_up :
            return self.alpha_arr_max[idx, 1]
        else:
            return self.alpha_arr_min[idx, 1]

    def update_power(self):
        self.alpha = np.array([self.calc_alpha(self.power[idx], self.p_up) for idx in range(len(self.power))])
        self.power = self.power - self.alpha*self.power*self.dz/1e-6
        self.power[self.power < 0] = 0

    def propgate(self, p_in):
        last_power = np.copy(self.power)
        self.power[1:] = self.power[:-1]
        self.power[0] = p_in
        # print(self.power)
        self.p_up = (p_in - self.last_p ) < 0
        # print(len(np.argwhere(self.p_up == True)))

    def update(self, p_in):
        self.propgate(p_in)
        self.update_power()

    def out(self,p_in):
        for i in range(len(self.power)):
            self.update(p_in)
        self.last_p = p_in
        return self.power[-1]
```

# Bistable ENZ device in Neural Networks

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from keras.utils.np_utils import to_categorical
from sklearn.metrics import confusion_matrix
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
class ENZ_act(tf.keras.layers.Layer):
    def __init__(self, **kwargs):
        super(ENZ_act, self).__init__(**kwargs)
        self.block_1 = 0
        self.block_2 = 2.4
        self.block_3 = 6.9
        self.slop_1 = 0
        self.slop_2 = 0.6
        self.slop_3 = 0.21
        self.slop_4 = 0.5
        self.intercept_3 = 0.95
        self.intercept_4 = -1.05
    
    def build(self, input_shape):
        return super().build(input_shape)
    
    def call(self, x):
        return tf.where(x<self.block_1, tf.multiply(x, self.slop_1), 
               tf.where(x<self.block_2, tf.multiply(x, self.slop_2),
               tf.where(x<self.block_3, tf.add(tf.multiply(x, self.slop_3), self.intercept_3),tf.add(tf.multiply(x, self.slop_4), self.intercept_4))))
```

```python
#plot of activation function derived from ENZ material behaviour
enz = ENZ_act()
a = tf.constant(np.arange(-10, 10, 0.1), dtype = tf.float32)
plt.plot(a, enz(a))
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()
```

```python
# Model / data parameters
num_classes = 10
input_shape = (28, 28, 1)

# Load the data and split it between train and test sets
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Scale images to the [0, 1] range
x_train = x_train.astype("float32") / 255
x_test = x_test.astype("float32") / 255
# Make sure images have shape (28, 28, 1)
train_data = np.expand_dims(x_train, -1)
test_data = np.expand_dims(x_test, -1)


# convert class vectors to binary class matrices
train_label = to_categorical(y_train, num_classes)
test_label = to_categorical(y_test, num_classes)
```

```python
#MNIST relu
act = 'relu'
epochs = 30
input_shape=(28,28,1)
model = tf.keras.Sequential(
    [
        tf.keras.Input(shape=input_shape),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(400,activation=act),
        tf.keras.layers.Dense(200,activation=act),

        tf.keras.layers.Dense(10,activation='softmax'),
    ]
)

model.summary()

batch_size = 128
callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=3)
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

history_MNIST_Relu=model.fit(train_data, train_label, batch_size=batch_size, epochs=epochs, validation_data=(test_data, test_label),verbose=0)



test_predictions = model.predict(test_data)
cm = confusion_matrix(y_true = np.argmax(test_label,axis=1), y_pred = np.argmax(test_predictions,axis=1))
sns.heatmap(cm,annot=True,fmt='d',linewidths=2,cmap="YlGnBu")
```
