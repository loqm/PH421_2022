$\large Abstract$

Optical parametric generation is treated as a particular case of parametric amplification of two waves at frequencies ùúî1 and ùúî2 from a pump wave ùúî3 propagating in a nonlinear crystal configuration in which initial intensities for ùúî1 and ùúî2 at the input plane of the material are equal to zero. This research is based on the theory of optical parametric fluorescence. According to the quantum-mechanical model, there is a probability for a photon ùúî3 to be spontaneously scattered into photons ‚Ñèùúî1 and ‚Ñèùúî2, respectively. We introduce a critical length over which the first signal and idler photons are created. This new approach allows us to take into account optical parametric fluorescence not only at the entrance of material but also over its entire interaction length. We show that this can widely modify the calculated generated intensities. This model is applied to a KTP optical parametric generator and amplifier pumped by the second harmonic (532 nm) of a Nd:YAG picosecond laser.

$\large Introduction $

Optical parametric oscillators generate a second-order nonlinear process that involves three optical frequencies: the pump, signal, and idler frequencies. An incident laser pump beam is first focused into a nonlinear crystal. The light field of the pump beam will generate multiple fields due to variances in the crystal‚Äôs refractive index. The generated field with higher frequency is referred to as the signal while the lower frequency field is called the idler.  The pump, signal, and idler waves interact such that the signal and idler waves amplify while the pump wave weakens. The created gain allows the resonant waves (either the signal, idler, or both) to oscillate within the resonator. The OPO is described as singly resonant when the cavity is matched to either the pump, signal or idler frequency. The system is doubly resonant when two are set to resonate together.Several authors developed a quantum-mechanical model that describes the first-generated photons as being emitted spontaneously from pump photons at the entrance plane of the crystal. This process is known as optical parametric fluorescence (OPF). Thus OPF provides the first photons that will take part in parametric amplification. Although OPF has been clearly described and observed experimentally, the number of noise photons‚Äîconsidered as the source for the parametric amplification‚Äîhas been the subject of different approaches based on assumptions that are not clearly justified physically. We have developed a simple model in which the relation between pump and noise photons is clearly defined. By taking into account the probability for a pump photon to be emitted spontaneously into a signal photon and an idler photon, we defined a critical length by analogy with laser amplification into a medium.
This is the length over which N photons produce one photon at each generated frequency. The two photons produced over that critical length are then taken as signal and idler input intensities for parametric amplification over the remaining distance in the crystal. Then we further developed our theoretical model by including OPF not only at the entrance plane of the crystal but also over the entire photon path in the crystal. We demonstrate that since noise
photons are not negligible in comparison to signal and idler intensities, OPF has to be considered in parametric interactions.

$\large Conclusion :$

We have investigated a new method for simulating optical parametric generation. We considered that initially generated photons were scattered by pump photons over a length defined as the critical length. This approach does not contain any other assumptions that are difficult to justify (such as, for instance, the number of initial noise photons that we have to consider as an initial parameter for parametric amplification) and requires only knowledge of the characteristics of the pump beam and the nonlinear crystal. Moreover, we have demonstrated that if optical parametric fluorescence is the physical phenomenon that gives rise to the initial signal and idler photons, it cannot be disregarded in parametric amplification. As evidence, a certain probability for a pump photon to scatter spontaneously into one photon in each mode exists over the whole length of interaction. We have shown that a part of the photons created by optical parametric fluorescence can widely enhance the total generated output energy as long as these photons are sufficiently amplified by pump pulses. The validity of the method was tested by comparison of the computational results with experimental data provided by a picosecond KTP optical parametric generator and amplifier. We observed a reasonably good agreement with experimental results. A more complete theoretical description of our picosecond optical parametric generator and amplifier is in progress. This method can be further developed to apply to different optical parametric devices and to consider special cases, such as the phase-mismatch case.
